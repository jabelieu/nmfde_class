```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python

np.random.seed(42)

# Parameters
n_samples = 100
true_slope = 2.0
true_intercept = 5.0
noise_std = 3.0  # standard deviation of noise

# Generate data
x = np.random.rand(n_samples) * 10
noise = np.random.randn(n_samples) * noise_std
t = true_slope * x + true_intercept + noise

# Reshape x for scikit-learn
X = x.reshape(-1, 1)


# Fit linear regression
model = LinearRegression()
model.fit(X, t)
# print("Coefficient:", model.coef_[0])
# print("Intercept:", model.intercept_)

# Predict
x_plot = np.linspace(x.min(), x.max(), 100).reshape(-1, 1)
t_pred = model.predict(x_plot)
fs=18
# Plot
plt.figure(figsize=(6,5),dpi=150)
plt.scatter(x, t, alpha=0.8, label='Data')
# plt.plot(x_plot, t_pred, color='red', label='Linear fit')
plt.xlabel(r'$\mathbf{\vec{x}}$',fontsize=fs)
plt.ylabel(r'y',fontsize=fs)
plt.tick_params(direction='in')
plt.grid(ls='--',alpha=0.5)
# plt.legend()
plt.show()

# Plot
plt.figure(figsize=(6,5),dpi=150)
plt.scatter(x, t, alpha=0.8, label='Data')
plt.plot(x_plot, t_pred, color='red', label='Linear fit')
plt.xlabel(r'$\mathbf{\vec{x}}$',fontsize=fs)
plt.ylabel(r'y',fontsize=fs)
plt.tick_params(direction='in')
plt.grid(ls='--',alpha=0.5)
# plt.legend()
plt.show()
```
