# Boundary Conditions for the 1D Wave Equation in First‚ÄìOrder Form

We begin with the 1D wave equation on a spatial interval $x \in [0,L]$:
$$
u_{tt} = c^2 u_{xx}.
$$

Introduce first‚Äìorder variables
$$
p = u_t, \qquad r = c\,u_x.
$$

Then the first‚Äìorder system is
$$
\begin{aligned}
u_t &= p,\\[2mm]
p_t - c\, r_x &= 0,\\[2mm]
r_t - c\, p_x &= 0.
\end{aligned}
$$

We will treat this as a **$2\times 2$ first-order hyperbolic system** for $(p,r)$. The variable $u$ is then reconstructed from $p$ and $r$.

---

## Matrix form and eigen‚Äìdecomposition

Collect the first-order variables into a vector
$$
q = 
\begin{pmatrix}
p \\[1mm] r
\end{pmatrix}.
$$

The PDEs for $p$ and $r$ may be written in the form
$$
q_t + A\,q_x = 0
$$
with
$$
A = 
\begin{pmatrix}
0 & -c \\[1mm]
-\,c & 0
\end{pmatrix}.
$$

We now compute the eigenvalues and eigenvectors of the matrix $A$.

### Eigenvalues

We solve
$$
\det(A - \lambda I) = 0.
$$

We have
$$
A - \lambda I =
\begin{pmatrix}
-\lambda & -c\\[1mm]
-\,c & -\lambda
\end{pmatrix},
\quad\Rightarrow\quad
\det(A - \lambda I) 
= \lambda^2 - c^2 = 0.
$$

Hence the eigenvalues are
$$
\lambda_\pm = \pm c.
$$

Thus:

- The **right‚Äìgoing characteristic** has speed $+c$.
- The **left‚Äìgoing characteristic** has speed $-c$.

### Eigenvectors

For $\lambda = +c$, we solve $(A - cI)v_+ = 0$:
$$
\begin{pmatrix}
- c & -c\\[1mm]
- c & -c
\end{pmatrix}
\begin{pmatrix} v_{+,1}\\[1mm] v_{+,2} \end{pmatrix}
= 0.
$$

One line is
$$
- c\,v_{+,1} - c\,v_{+,2} = 0 \quad\Rightarrow\quad v_{+,1} + v_{+,2} = 0,
$$
so an eigenvector is
$$
v_+ = \begin{pmatrix} 1 \\[1mm] -1 \end{pmatrix}.
$$

For $\lambda = -c$, we solve $(A + cI)v_- = 0$:
$$
\begin{pmatrix}
c & -c\\[1mm]
- c & c
\end{pmatrix}
\begin{pmatrix} v_{-,1}\\[1mm] v_{-,2} \end{pmatrix}
= 0.
$$

One line is
$$
c\,v_{-,1} - c\,v_{-,2} = 0 \quad\Rightarrow\quad v_{-,1} = v_{-,2},
$$
so an eigenvector is
$$
v_- = \begin{pmatrix} 1 \\[1mm] 1 \end{pmatrix}.
$$

We can assemble the eigenvector matrix
$$
S = 
\begin{pmatrix}
1 & 1\\[1mm]
-1 & 1
\end{pmatrix}
= \big[\,v_+ \;\; v_-\,\big],
$$
so that
$$
A = S\,\Lambda\,S^{-1}, \qquad 
\Lambda = 
\begin{pmatrix}
c & 0\\[1mm]
0 & -c
\end{pmatrix}.
$$

# Eigen-Decomposition and Characteristic Variables

We write the first-order wave system in the form  
$$
q_t + A\,q_x = 0,
\qquad
q = \begin{pmatrix} p \\ r \end{pmatrix}.
$$

The coefficient matrix is  
$$
A =
\begin{pmatrix}
0 & -c \\
-c & 0
\end{pmatrix}.
$$

---

## Eigenvalues

Compute the eigenvalues from  
$$
\det(A - \lambda I) = 0.
$$

Since  
$$
\det
\begin{pmatrix}
-\lambda & -c \\
-c & -\lambda
\end{pmatrix}
= \lambda^2 - c^2,
$$
we obtain  
$$
\lambda_+ = c,
\qquad
\lambda_- = -c.
$$

---

## Eigenvectors and the matrix $R$

For $\lambda_+ = c$, solve $(A - cI)v_+ = 0$:
$$
v_+ = \begin{pmatrix} 1 \\ -1 \end{pmatrix}.
$$

For $\lambda_- = -c$, solve $(A + cI)v_- = 0$:
$$
v_- = \begin{pmatrix} 1 \\ 1 \end{pmatrix}.
$$

Assemble the eigenvectors into the matrix  
$$
R =
\begin{pmatrix}
1 & 1 \\
-1 & 1
\end{pmatrix},
$$

and define  
$$
\Lambda =
\begin{pmatrix}
c & 0 \\
0 & -c
\end{pmatrix}.
$$

Thus  
$$
A = R \Lambda R^{-1}.
$$

---

## Compute $R^{-1}$

The determinant of $R$ is  
$$
\det(R) = 1\cdot 1 - (1)(-1) = 2.
$$

Thus  
$$
R^{-1}
= \frac{1}{2}
\begin{pmatrix}
1 & -1 \\
1 & \;\;1
\end{pmatrix}.
$$

---

## Characteristic variables

Define  
$$
w = R^{-1} q.
$$

Since  
$$
q = \begin{pmatrix} p \\ r \end{pmatrix},
\qquad
R^{-1} = \frac12
\begin{pmatrix}
1 & -1 \\
1 & 1
\end{pmatrix},
$$
we compute  
$$
w =
\begin{pmatrix}
w_- \\
w_+
\end{pmatrix}
=
\frac12
\begin{pmatrix}
p - r \\
p + r
\end{pmatrix}.
$$

Equivalently, if we choose the unscaled characteristic variables,
$$
\tilde w_+ = p + r, 
\qquad 
\tilde w_- = p - r,
$$
then  
$$
w_+ = \frac{\tilde w_+}{2}, 
\qquad 
w_- = \frac{\tilde w_-}{2}.
$$

---

## Expressing $(p,r)$ in terms of $(w_+, w_-)$

Because $q = Rw$, we have  
$$
\begin{pmatrix}
p \\
r
\end{pmatrix}
=
\begin{pmatrix}
1 & 1 \\
-1 & 1
\end{pmatrix}
\begin{pmatrix}
w_- \\
w_+
\end{pmatrix}.
$$

Thus  
$$
p = w_- + w_+,
\qquad
r = -w_- + w_+.
$$

Or in terms of $\tilde w_\pm$,
$$
p = \frac{\tilde w_+ + \tilde w_-}{2},
\qquad
r = \frac{\tilde w_+ - \tilde w_-}{2}.
$$

---

## Decoupled characteristic PDEs

Since  
$$
w = R^{-1} q,
\qquad
q = Rw,
$$
the PDE becomes  
$$
w_t + \Lambda w_x = 0.
$$

That is,
$$
(w_+)_t + c (w_+)_x = 0,
$$

$$
(w_-)_t - c (w_-)_x = 0.
$$

These are right- and left-moving waves, respectively.

---

## Direct (physical) boundary conditions

There are several standard types of boundary conditions for the wave equation written directly in terms of $u$ (and hence $p$ and $r$):

- **Dirichlet (displacement)**: specify the displacement
  $$
  u(0,t) = g_0(t), \qquad u(L,t) = g_L(t).
  $$

- **Neumann (flux / slope)**: specify the normal derivative
  $$
  u_x(0,t) = h_0(t), \qquad u_x(L,t) = h_L(t).
  $$

- **Periodic**: identify the ends of the domain
  $$
  u(0,t) = u(L,t), \qquad u_x(0,t) = u_x(L,t).
  $$

- **Outflow / non-reflecting**: approximate that waves leave the domain with no incoming signal from outside
  $$
  u_t(0,t) + c\,u_x(0,t) = g_{\text{left}}(t), \qquad
  u_t(L,t) - c\,u_x(L,t) = g_{\text{right}}(t),
  $$
  where typically $g_{\text{left}}$ and $g_{\text{right}}$ are taken to be zero (no incoming waves).

The first-order variables are related by
$$
p = u_t, \qquad r = c\,u_x.
$$

Thus any boundary condition on $u$ and $u_x$ induces constraints on $p$ and $r$, and in turn on $w_+$ and $w_-$.

## Periodic boundary conditions

For **periodic boundary conditions**, we identify the ends of the interval:
$$
u(0,t) = u(L,t), \qquad u_x(0,t) = u_x(L,t), \quad \forall t.
$$

In terms of the first-order variables:
$$
p(0,t) = p(L,t), \qquad r(0,t) = r(L,t).
$$

Because the characteristic variables are linear combinations of $p$ and $r$,
$$
w_+(0,t) = p(0,t) + r(0,t) = p(L,t) + r(L,t) = w_+(L,t),
$$
$$
w_-(0,t) = p(0,t) - r(0,t) = p(L,t) - r(L,t) = w_-(L,t).
$$

So **periodic BC in characteristic form** is simply:
$$
w_\pm(0,t) = w_\pm(L,t).
$$

In a finite-difference code, this translates into:

- ghost cells at the left boundary are copied from the right interior,
- ghost cells at the right boundary are copied from the left interior,

for either $(p,r)$ or $(w_+,w_-)$.


```python
import numpy as np
import matplotlib.pyplot as plt

##########################################
# 1D wave equation in first-order form
#   u_tt = c^2 u_xx
# Introduce:
#   p = u_t
#   r = c u_x
#
# First-order system (for [p, r]):
#   p_t - c r_x = 0
#   r_t - c p_x = 0
#
# In vector form w = [p, r]^T:
#   w_t + A w_x = 0
# with
#   A = [[ 0, -c],
#        [-c,  0]]
#
# Characteristic decomposition:
#   Eigenvalues: Œª_- = -c (left-going), Œª_+ = +c (right-going)
#   Eigenvectors:
#     v_- = [1,  1]^T  (for Œª=-c)
#     v_+ = [1, -1]^T  (for Œª=+c)
#
# Define characteristic variables:
#   w_- = 0.5*(p + r)   (left-going, speed -c)
#   w_+ = 0.5*(p - r)   (right-going, speed +c)
#
# Reconstruction:
#   p = w_- + w_+
#   r = w_- - w_+
#
# We will:
#   1. Use ghost cells (2 on each side) for Beam‚ÄìWarming.
#   2. Enforce **periodic BCs** via ghost cells:
#         q[1]   <- q[Nx+1]
#         q[0]   <- q[Nx]
#         q[Nx+2] <- q[2]
#         q[Nx+3] <- q[3]
#   3. Transform [p, r] -> [w_-, w_+].
#   4. Apply BW update to each characteristic variable separately.
#   5. Transform back to [p, r].
#   6. Update u using u_t = p.
##########################################

##########################################
# Beam‚ÄìWarming scalar stencils (with ghost cells)
# We assume:
#   - ghost cells are already filled
#   - interior indices are i = 2 ... Nx+1
#   - total allocated size is Ntot = Nx + 4
##########################################
def BW_derivatives(q, a, dx, i_start, i_end):
    """
    Compute first and second upwind derivatives for Beam‚ÄìWarming
    for scalar q on indices i_start...i_end (inclusive),
    assuming ghost cells outside that range are valid.

    q : 1D array with ghost cells
    a : scalar advection speed
    dx: grid spacing
    i_start, i_end: integer indices of interior cells (e.g., 2, Nx+1)

    Returns:
        D1, D2 (arrays same shape as q, values set only on interior indices)
    """
    D1 = np.zeros_like(q)
    D2 = np.zeros_like(q)

    if a == 0.0:
        return D1, D2

    if a > 0:
        # Backward-biased (i, i-1, i-2)
        i = np.arange(i_start, i_end+1)
        D1[i] = (3.0*q[i] - 4.0*q[i-1] + q[i-2]) / (2.0*dx)
        D2[i] = (q[i] - 2.0*q[i-1] + q[i-2]) / (dx*dx)
    else:
        # Forward-biased (i, i+1, i+2)
        i = np.arange(i_start, i_end+1)
        D1[i] = (-3.0*q[i] + 4.0*q[i+1] - q[i+2]) / (2.0*dx)
        D2[i] = (q[i] - 2.0*q[i+1] + q[i+2]) / (dx*dx)

    return D1, D2

def BW_timestep_scalar(q, a, dx, dt, i_start, i_end):
    """
    One Beam‚ÄìWarming step for q_t + a q_x = 0 on interior cells.
    Uses the Taylor expansion:
       q^{n+1} = q^n - a dt D1 + 0.5 (a dt)^2 D2
    """
    D1, D2 = BW_derivatives(q, a, dx, i_start, i_end)
    q_new = q.copy()
    q_new[i_start:i_end+1] = (
        q[i_start:i_end+1]
        - a*dt*D1[i_start:i_end+1]
        + 0.5*(a*dt)*(a*dt)*D2[i_start:i_end+1]
    )
    return q_new

##########################################
# Periodic boundary conditions via ghost cells
#
# Layout of q:
#   indices: 0,1 | 2,...,Nx+1 | Nx+2,Nx+3
#   interior (physical): 2 ... Nx+1  (Nx cells)
#
# Periodicity:
#   left ghosts copy from right interior,
#   right ghosts copy from left interior.
##########################################
def apply_periodic(q, Nx):
    """
    Fill 2 ghost cells on each side of q with periodic values.

    Layout:
        0,1 | 2,...,Nx+1 | Nx+2,Nx+3

    Periodic mapping:
        q[1]    <- q[Nx+1]   (last interior cell)
        q[0]    <- q[Nx]     (second-to-last interior)
        q[Nx+2] <- q[2]      (first interior)
        q[Nx+3] <- q[3]      (second interior)
    """
    q[1]      = q[Nx+1]
    q[0]      = q[Nx]
    q[Nx+2]   = q[2]
    q[Nx+3]   = q[3]

##########################################
# Main solver for 1D wave equation (first-order form, periodic)
##########################################
def solve_wave_1d_BW_char_periodic(
    c=1.0,
    L=1.0,
    Nx=200,
    Tfinal=1.0,
    CFL=0.5
):
    dx = L / Nx
    # time step from CFL: use |Œª_max| = c
    dt = CFL * dx / c
    Nt = int(np.ceil(Tfinal / dt))
    dt = Tfinal / Nt  # snap to hit Tfinal exactly

    # Total size with 2 ghost cells on each side
    Ntot = Nx + 4
    i_start, i_end = 2, Nx+1  # interior indices

    # Allocate fields with ghost cells
    u = np.zeros(Ntot)
    p = np.zeros(Ntot)
    r = np.zeros(Ntot)

    # Coordinates at cell centers (for plotting)
    # Interior cells i=2..Nx+1 cover [0,L)
    x = (np.arange(Ntot) - 2 + 0.5) * dx

    # Initial condition: a Gaussian bump in u, zero initial p
    x_phys = x[i_start:i_end+1]
    u[i_start:i_end+1] = np.exp(-100.0 * (x_phys - 0.5*L)**2)
    p[i_start:i_end+1] = 0.0

    # Enforce periodic BC on u and p before computing r
    apply_periodic(u, Nx)
    apply_periodic(p, Nx)

    # Initialize r = c u_x using centered difference (uses ghost cells)
    r[i_start:i_end+1] = c * (u[i_start+1:i_end+2] - u[i_start-1:i_end]) / (2.0*dx)

    # Time stepping
    for n in range(Nt):
        t = n*dt  # not really used, but kept for completeness

        # 1) Enforce periodic BC on u, p, r
        apply_periodic(u, Nx)
        apply_periodic(p, Nx)
        apply_periodic(r, Nx)

        # 2) Transform (p, r) into characteristic variables:
        #    w_- = 0.5*(p + r), w_+ = 0.5*(p - r)
        w_minus = 0.5*(p + r)  # left-going, speed -c
        w_plus  = 0.5*(p - r)  # right-going, speed +c

        # Enforce periodic BC on characteristic variables as well
        apply_periodic(w_minus, Nx)
        apply_periodic(w_plus,  Nx)

        # 3) Beam‚ÄìWarming update on each characteristic field
        #    w_-: w_-_t + (-c) w_-_x = 0  (a = -c)
        #    w_+: w_+_t + (+c) w_+_x = 0  (a = +c)
        w_minus_new = BW_timestep_scalar(w_minus, a=-c, dx=dx, dt=dt,
                                         i_start=i_start, i_end=i_end)
        w_plus_new  = BW_timestep_scalar(w_plus,  a=+c, dx=dx, dt=dt,
                                         i_start=i_start, i_end=i_end)

        # 4) Transform back to p, r
        p = w_minus_new + w_plus_new
        r = w_minus_new - w_plus_new

        # 5) Update u with u_t = p (forward Euler)
        u[i_start:i_end+1] += dt * p[i_start:i_end+1]

        # (Optional) re-apply periodic to u now or at next step
        # We'll do it at the top of the loop.

    # Extract physical interior for plotting and output
    x_phys = x[i_start:i_end+1]
    u_phys = u[i_start:i_end+1]
    p_phys = p[i_start:i_end+1]
    r_phys = r[i_start:i_end+1]
    return x_phys, u_phys, p_phys, r_phys

##########################################
# Run a demo
##########################################
if __name__ == "__main__":
    c = 1.0
    L = 1.0
    Nx = 200
    Tfinal = 1.0
    CFL = 0.8

    x, u, p, r = solve_wave_1d_BW_char_periodic(c=c, L=L, Nx=Nx, Tfinal=Tfinal, CFL=CFL)

    plt.figure(figsize=(6,4))
    plt.plot(x, u, label='u(x,T)')
    plt.xlabel('x'); plt.ylabel('u')
    plt.title('1D Wave Equation ‚Äì Beam‚ÄìWarming + Characteristics (Periodic BC)')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

# Dirichlet
---

## Dirichlet boundary condition for the displacement $u$

A **Dirichlet boundary condition** for the wave equation is
$$
u(0,t) = g_0(t), \qquad u(L,t) = g_L(t).
$$

These prescribe **the values of the displacement $u$** at the left and right boundaries.

### What about $p$ and $r$?

Since
$$
p = u_t, \qquad r = c\,u_x,
$$
we obtain:

- At $x=0$:
  $$
  p(0,t) = g_0'(t), \qquad r(0,t) = c\,u_x(0,t).
  $$
  The derivative $u_x(0,t)$ must be obtained numerically (via ghost points or one-sided finite differences).

- At $x=L$:
  $$
  p(L,t) = g_L'(t), \qquad r(L,t) = c\,u_x(L,t).
  $$

Thus Dirichlet BCs for $u$ **do not directly prescribe** $r=c u_x$, but $r$ is constrained by the BC for $u$ once a spatial discretization is chosen.


---

## Characteristic form of the Dirichlet boundary condition

A hyperbolic system requires **only as many boundary conditions as incoming characteristic fields**:

- At **left boundary $x=0$**:  
  The **right-going** characteristic $w_+$ enters the domain (speed $+c$).  
  Therefore we must specify
  $$
  w_+(0,t) = p(0,t) + r(0,t).
  $$

- At **right boundary $x=L$**:  
  The **left-going** characteristic $w_-$ enters (speed $-c$).  
  Therefore we must specify
  $$
  w_-(L,t) = p(L,t) - r(L,t).
  $$

A Dirichlet boundary condition for $u$ supplies the necessary information, because
$$
u_t = p \quad\Rightarrow\quad p = \frac{\partial u}{\partial t} = g'(t),
$$
and then $r = c u_x$ is obtained from the spatial stencil enforcing the boundary.

Thus:

- At $x=0$:
  $$
  w_+(0,t) = g_0'(t) + c\,u_x(0,t).
  $$
- At $x=L$:
  $$
  w_-(L,t) = g_L'(t) - c\,u_x(L,t).
  $$

In numerical practice, the Dirichlet boundary condition for $u$ is enforced first, and the corresponding characteristic values $w_+$ and $w_-$ are set consistently.

```python
import numpy as np
import matplotlib.pyplot as plt

##########################################
# 1D wave equation in first-order form
#   u_tt = c^2 u_xx
# Introduce:
#   p = u_t
#   r = c u_x
#
# First-order system (for [p, r]):
#   p_t - c r_x = 0
#   r_t - c p_x = 0
#
# Characteristic variables:
#   w_- = 0.5*(p + r)   (left-going, speed -c)
#   w_+ = 0.5*(p - r)   (right-going, speed +c)
#
# Reconstruction:
#   p = w_- + w_+
#   r = w_- - w_+
#
# Here we impose *direct Dirichlet BCs* on u:
#   u(0,t) = gL(t), u(L,t) = gR(t)
# and use ghost cells for Beam‚ÄìWarming stencils.
##########################################

##########################################
# Beam‚ÄìWarming scalar stencils (with ghost cells)
##########################################
def BW_derivatives(q, a, dx, i_start, i_end):
    """
    Compute first and second upwind derivatives for Beam‚ÄìWarming
    for scalar q on indices i_start...i_end (inclusive),
    assuming ghost cells outside that range are valid.

    q : 1D array with ghost cells
    a : scalar advection speed
    dx: grid spacing
    i_start, i_end: integer indices of interior cells (e.g., 2, Nx+1)

    Returns:
        D1, D2 (arrays same shape as q, values set only on interior indices)
    """
    D1 = np.zeros_like(q)
    D2 = np.zeros_like(q)

    if a == 0.0:
        return D1, D2

    if a > 0:
        # Backward-biased (i, i-1, i-2)
        i = np.arange(i_start, i_end+1)
        D1[i] = (3.0*q[i] - 4.0*q[i-1] + q[i-2]) / (2.0*dx)
        D2[i] = (q[i] - 2.0*q[i-1] + q[i-2]) / (dx*dx)
    else:
        # Forward-biased (i, i+1, i+2)
        i = np.arange(i_start, i_end+1)
        D1[i] = (-3.0*q[i] + 4.0*q[i+1] - q[i+2]) / (2.0*dx)
        D2[i] = (q[i] - 2.0*q[i+1] + q[i+2]) / (dx*dx)

    return D1, D2

def BW_timestep_scalar(q, a, dx, dt, i_start, i_end):
    """
    One Beam‚ÄìWarming step for q_t + a q_x = 0 on interior cells.
    Uses the Taylor expansion:
       q^{n+1} = q^n - a dt D1 + 0.5 (a dt)^2 D2
    """
    D1, D2 = BW_derivatives(q, a, dx, i_start, i_end)
    q_new = q.copy()
    q_new[i_start:i_end+1] = (
        q[i_start:i_end+1]
        - a*dt*D1[i_start:i_end+1]
        + 0.5*(a*dt)*(a*dt)*D2[i_start:i_end+1]
    )
    return q_new

##########################################
# Direct Dirichlet BCs on u with ghost cells
#
# Layout of arrays:
#   indices: 0,1 | 2,...,Nx+1 | Nx+2,Nx+3
#   interior (physical): 2 ... Nx+1  (Nx cells)
#
# We impose:
#   u(0,t) = gL(t),
#   u(L,t) = gR(t),
# and use *linear extrapolation* for ghost cells:
#   u_{-1} ‚âà 2 u_0 - u_1, u_{-2} ‚âà 2 u_{-1} - u_0, etc.
##########################################
def apply_dirichlet_u(u, Nx, dx, t, gL_func, gR_func):
    """
    Apply Dirichlet BCs to u at left/right via interior + ghost cells.

    Ghost layout:
      indices: 0,1 | 2,...,Nx+1 | Nx+2,Nx+3
      physical interior: 2 ... Nx+1
      left boundary at x=0 approximated by cell i=2
      right boundary at x=L approximated by cell i=Nx+1

    gL_func, gR_func: functions of time t returning u(0,t), u(L,t)
    """
    gL = gL_func(t)
    gR = gR_func(t)

    # Set boundary values on interior boundary cells
    u[2]      = gL    # left boundary value
    u[Nx+1]   = gR    # right boundary value

    # Linear extrapolation to ghost cells for u
    # Left side: use u[2] (gL) and u[3]
    u[1]      = 2.0*gL - u[3]
    u[0]      = 2.0*gL - u[4]

    # Right side: use u[Nx+1] (gR) and u[Nx]
    u[Nx+2]   = 2.0*gR - u[Nx]
    u[Nx+3]   = 2.0*gR - u[Nx-1]

##########################################
# Dirichlet BCs on p = u_t
#
# If u(0,t) = gL(t), then p(0,t) = gL'(t),
# similarly at x=L with gR'(t).
#
# We do the same ghost strategy as for u.
##########################################
def apply_dirichlet_p(p, Nx, t, gL_t_func, gR_t_func):
    """
    Apply Dirichlet BCs for p = u_t from gL'(t), gR'(t),
    and fill ghost cells by linear extrapolation.
    """
    pL = gL_t_func(t)
    pR = gR_t_func(t)

    p[2]      = pL
    p[Nx+1]   = pR

    # Extrapolate ghost cells for p
    p[1]      = 2.0*pL - p[3]
    p[0]      = 2.0*pL - p[4]

    p[Nx+2]   = 2.0*pR - p[Nx]
    p[Nx+3]   = 2.0*pR - p[Nx-1]

##########################################
# Main solver for 1D wave equation (first-order form)
# with *direct Dirichlet BCs on u*
##########################################
def solve_wave_1d_BW_char_dirichlet(
    c=1.0,
    L=1.0,
    Nx=200,
    Tfinal=1.0,
    CFL=0.5,
    gL=lambda t: 0.0,
    gR=lambda t: 0.0,
    gL_t=lambda t: 0.0,
    gR_t=lambda t: 0.0
):
    dx = L / Nx
    # time step from CFL: use |Œª_max| = c
    dt = CFL * dx / c
    Nt = int(np.ceil(Tfinal / dt))
    dt = Tfinal / Nt  # snap to hit Tfinal exactly

    # Total size with 2 ghost cells on each side
    Ntot = Nx + 4
    i_start, i_end = 2, Nx+1  # interior indices

    # Allocate fields with ghost cells
    u = np.zeros(Ntot)
    p = np.zeros(Ntot)
    r = np.zeros(Ntot)

    # Coordinates at cell centers (for plotting)
    # Interior cells i=2..Nx+1 cover [0,L]
    x = (np.arange(Ntot) - 2 + 0.5) * dx

    # Initial condition: e.g., a Gaussian bump for u, zero initial p
    x_phys = x[i_start:i_end+1]
    u[i_start:i_end+1] = np.exp(-100.0 * (x_phys - 0.5*L)**2)
    p[i_start:i_end+1] = 0.0

    # Enforce Dirichlet BC on u and p at t=0
    apply_dirichlet_u(u, Nx, dx, 0.0, gL, gR)
    apply_dirichlet_p(p, Nx, 0.0, gL_t, gR_t)

    # Initialize r = c u_x using centered difference (uses ghost cells of u)
    r[i_start:i_end+1] = c * (u[i_start+1:i_end+2] - u[i_start-1:i_end]) / (2.0*dx)

    # Time stepping
    for n in range(Nt):
        t = n*dt

        # 1) Enforce Dirichlet BCs on u and p for current time
        apply_dirichlet_u(u, Nx, dx, t,   gL, gR)
        apply_dirichlet_p(p, Nx,    t,   gL_t, gR_t)

        # 2) Recompute r = c u_x with current u (and its ghosts)
        r[i_start:i_end+1] = c * (u[i_start+1:i_end+2] - u[i_start-1:i_end]) / (2.0*dx)

        # 3) Transform (p, r) into characteristic variables:
        #    w_- = 0.5*(p + r), w_+ = 0.5*(p - r)
        w_minus = 0.5*(p + r)  # left-going, speed -c
        w_plus  = 0.5*(p - r)  # right-going, speed +c

        # NOTE: we do NOT impose characteristic BCs explicitly here;
        # ghost values for w_- and w_+ are inherited from p and r,
        # which came from direct BCs on u and p + the derivative r=c u_x.

        # 4) Beam‚ÄìWarming update on each characteristic field
        #    w_-: w_-_t + (-c) w_-_x = 0  (a = -c)
        #    w_+: w_+_t + (+c) w_+_x = 0  (a = +c)
        w_minus_new = BW_timestep_scalar(w_minus, a=-c, dx=dx, dt=dt,
                                         i_start=i_start, i_end=i_end)
        w_plus_new  = BW_timestep_scalar(w_plus,  a=+c, dx=dx, dt=dt,
                                         i_start=i_start, i_end=i_end)

        # 5) Transform back to p, r
        p = w_minus_new + w_plus_new
        r = w_minus_new - w_plus_new

        # 6) Update u with u_t = p (forward Euler in time)
        u[i_start:i_end+1] += dt * p[i_start:i_end+1]

        # 7) Re-enforce Dirichlet on u at t+dt (for the *next* step)
        apply_dirichlet_u(u, Nx, dx, t+dt, gL, gR)

    # Extract physical interior for plotting and output
    x_phys = x[i_start:i_end+1]
    u_phys = u[i_start:i_end+1]
    p_phys = p[i_start:i_end+1]
    r_phys = r[i_start:i_end+1]
    return x_phys, u_phys, p_phys, r_phys

##########################################
# Run a demo with homogeneous Dirichlet:
#   u(0,t) = u(L,t) = 0
##########################################
if __name__ == "__main__":
    c = 1.0
    L = 1.0
    Nx = 200
    Tfinal = 1.0
    CFL = 0.1

    # Homogeneous Dirichlet: gL = gR = 0, and derivatives are 0 as well
    x, u, p, r = solve_wave_1d_BW_char_dirichlet(
        c=c, L=L, Nx=Nx, Tfinal=Tfinal, CFL=CFL,
        gL=lambda t: 0.0,
        gR=lambda t: 0.0,
        gL_t=lambda t: 0.0,
        gR_t=lambda t: 0.0
    )

    plt.figure(figsize=(6,4))
    plt.plot(x, u, label='u(x,T)')
    plt.xlabel('x'); plt.ylabel('u')
    plt.title('1D Wave Equation ‚Äì Beam‚ÄìWarming + Direct Dirichlet BCs')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
###################################
#
# A seconed order bounadry conditon  
#
# drop this in above and you will see 
# improved solutions with a bigger CFL
#
def apply_dirichlet_u(u, Nx, dx, t, gL_func, gR_func):
    """
    Apply *second-order accurate* Dirichlet BCs to u at left/right via ghost cells.

    Ghost layout:
      indices: 0,1 | 2,...,Nx+1 | Nx+2,Nx+3
      physical interior: 2 ... Nx+1
      left boundary corresponds to u[2]
      right boundary corresponds to u[Nx+1]

    Uses *quadratic (three-point) extrapolation* for ghost cells:
        u[-1] = 3 u0 - 3 u1 + u2
        u[-2] = 6 u0 - 8 u1 + 3 u2

    where (u0,u1,u2) = (u[2],u[3],u[4]) on the left and
                       (u[Nx+1],u[Nx],u[Nx-1]) on the right.
    """
    gL = gL_func(t)
    gR = gR_func(t)

    # ---------------------------------------------
    # LEFT BOUNDARY: u(0,t) = gL(t)
    # ---------------------------------------------
    # Set interior boundary point exactly
    u[2] = gL

    # Quadratic extrapolation for ghost cells:
    # u[1] = u(x=-dx)   ‚Üê u_-1
    # u[0] = u(x=-2dx)  ‚Üê u_-2
    #
    # Using: (u0,u1,u2) = (u[2],u[3],u[4])
    u0 = u[2]
    u1 = u[3]
    u2 = u[4]

    u[1] = 3*u0 - 3*u1 + u2             # first ghost (x = -dx)
    u[0] = 6*u0 - 8*u1 + 3*u2           # second ghost (x = -2dx)

    # ---------------------------------------------
    # RIGHT BOUNDARY: u(L,t) = gR(t)
    # ---------------------------------------------
    u[Nx+1] = gR

    # Quadratic extrapolation on the right:
    # Using reversed stencil (mirror symmetry)
    # Points: u0 = u[Nx+1], u1 = u[Nx], u2 = u[Nx-1]
    u0 = u[Nx+1]
    u1 = u[Nx]
    u2 = u[Nx-1]

    u[Nx+2] = 3*u0 - 3*u1 + u2          # first ghost at right
    u[Nx+3] = 6*u0 - 8*u1 + 3*u2        # second ghost at right
```

## Neumann boundary conditions

A **Neumann boundary condition** for the wave equation prescribes the normal derivative:
$$
u_x(0,t) = h_0(t), \qquad u_x(L,t) = h_L(t).
$$

In terms of the first-order variables:
$$
r(0,t) = c\,u_x(0,t) = c\,h_0(t) =: r_0(t),
$$
$$
r(L,t) = c\,u_x(L,t) = c\,h_L(t) =: r_L(t).
$$

We do **not** prescribe $p$ directly; instead we prescribe the ‚Äúflux‚Äù $r$.

### Characteristic form at $x=0$

Recall
$$
w_+ = p + r, \qquad w_- = p - r.
$$

At $x=0$, the Neumann condition gives
$$
r(0,t) = r_0(t),
$$
so
$$
\frac{w_+(0,t) - w_-(0,t)}{2} = r_0(t)
\quad\Rightarrow\quad
w_+(0,t) - w_-(0,t) = 2\,r_0(t).
$$

At the **left boundary $x=0$**:

- incoming characteristic: $w_+$,
- outgoing characteristic: $w_-$.

We therefore express the **incoming** characteristic in terms of the **outgoing** one and the Neumann data:
$$
\boxed{
w_+(0,t) = w_-(0,t) + 2\,r_0(t).
}
$$

Special case: **homogeneous Neumann** ($u_x(0,t)=0$, so $r_0(t)=0$) gives
$$
w_+(0,t) = w_-(0,t),
$$
i.e. reflection with no sign change (a ‚Äúfree‚Äù end for a string).

### Characteristic form at $x=L$

At $x=L$, we have
$$
r(L,t) = r_L(t)
\quad\Rightarrow\quad
\frac{w_+(L,t) - w_-(L,t)}{2} = r_L(t)
\quad\Rightarrow\quad
w_+(L,t) - w_-(L,t) = 2\,r_L(t).
$$

Now at the **right boundary $x=L$**:

- incoming characteristic: $w_-$,
- outgoing characteristic: $w_+$.

So we solve for the incoming field:
$$
\boxed{
w_-(L,t) = w_+(L,t) - 2\,r_L(t).
}
$$

Again, for **homogeneous Neumann** ($r_L(t)=0$):
$$
w_-(L,t) = w_+(L,t),
$$
free-end reflection.

```python
import numpy as np
import matplotlib.pyplot as plt

##########################################
# 1D wave equation in first-order form
#   u_tt = c^2 u_xx
# Introduce:
#   p = u_t
#   r = c u_x
#
# First-order system (for [p, r]):
#   p_t - c r_x = 0
#   r_t - c p_x = 0
#
# Characteristic variables:
#   w_- = 0.5*(p + r)   (left-going, speed -c)
#   w_+ = 0.5*(p - r)   (right-going, speed +c)
#
# Reconstruction:
#   p = w_- + w_+
#   r = w_- - w_+
#
# Here we impose *direct Neumann BCs* on u:
#   u_x(0,t) = hL(t), u_x(L,t) = hR(t)
# via ghost cells for u, and a simple Neumann-like treatment for p.
##########################################

##########################################
# Beam‚ÄìWarming scalar stencils (with ghost cells)
##########################################
def BW_derivatives(q, a, dx, i_start, i_end):
    """
    Compute first and second upwind derivatives for Beam‚ÄìWarming
    for scalar q on indices i_start...i_end (inclusive),
    assuming ghost cells outside that range are valid.

    q : 1D array with ghost cells
    a : scalar advection speed
    dx: grid spacing
    i_start, i_end: integer indices of interior cells (e.g., 2, Nx+1)

    Returns:
        D1, D2 (arrays same shape as q, values set only on interior indices)
    """
    D1 = np.zeros_like(q)
    D2 = np.zeros_like(q)

    if a == 0.0:
        return D1, D2

    if a > 0:
        # Backward-biased (i, i-1, i-2)
        i = np.arange(i_start, i_end+1)
        D1[i] = (3.0*q[i] - 4.0*q[i-1] + q[i-2]) / (2.0*dx)
        D2[i] = (q[i] - 2.0*q[i-1] + q[i-2]) / (dx*dx)
    else:
        # Forward-biased (i, i+1, i+2)
        i = np.arange(i_start, i_end+1)
        D1[i] = (-3.0*q[i] + 4.0*q[i+1] - q[i+2]) / (2.0*dx)
        D2[i] = (q[i] - 2.0*q[i+1] + q[i+2]) / (dx*dx)

    return D1, D2

def BW_timestep_scalar(q, a, dx, dt, i_start, i_end):
    """
    One Beam‚ÄìWarming step for q_t + a q_x = 0 on interior cells.
    Uses the Taylor expansion:
       q^{n+1} = q^n - a dt D1 + 0.5 (a dt)^2 D2
    """
    D1, D2 = BW_derivatives(q, a, dx, i_start, i_end)
    q_new = q.copy()
    q_new[i_start:i_end+1] = (
        q[i_start:i_end+1]
        - a*dt*D1[i_start:i_end+1]
        + 0.5*(a*dt)*(a*dt)*D2[i_start:i_end+1]
    )
    return q_new

##########################################
# Direct Neumann BCs on u with ghost cells
#
# Layout of arrays:
#   indices: 0,1 | 2,...,Nx+1 | Nx+2,Nx+3
#   interior (physical): 2 ... Nx+1  (Nx cells)
#
# We impose:
#   u_x(0,t) = hL(t),
#   u_x(L,t) = hR(t),
#
# Approximate derivative at the boundary using the face between
# the first ghost and first interior cell:
#   (u[2] - u[1]) / dx ‚âà hL(t)   => u[1]  = u[2] - dx hL
#   (u[Nx+2] - u[Nx+1]) / dx ‚âà hR(t) => u[Nx+2] = u[Nx+1] + dx hR
#
# Then linearly extrapolate one more ghost:
#   u[0]      = 2 u[1]      - u[2]
#   u[Nx+3]   = 2 u[Nx+2]   - u[Nx+1]
##########################################
def apply_neumann_u(u, Nx, dx, t, hL_func, hR_func):
    """
    Apply Neumann BCs to u at left/right via ghost cells.

    hL_func, hR_func: functions of time t returning u_x(0,t), u_x(L,t)

    Ghost layout:
      indices: 0,1 | 2,...,Nx+1 | Nx+2,Nx+3
      physical interior: 2 ... Nx+1
    """
    hL = hL_func(t)
    hR = hR_func(t)

    # Left boundary: (u[2] - u[1]) / dx = hL
    u[1] = u[2] - dx * hL
    u[0] = 2.0 * u[1] - u[2]

    # Right boundary: (u[Nx+2] - u[Nx+1]) / dx = hR
    u[Nx+2] = u[Nx+1] + dx * hR
    u[Nx+3] = 2.0 * u[Nx+2] - u[Nx+1]

##########################################
# "Neumann-like" BCs on p = u_t
#
# The Neumann condition is specified for u, not p.
# A simple choice is to impose zero derivative on p at the boundary:
#   p_x(0,t) ‚âà 0, p_x(L,t) ‚âà 0
#
# This corresponds to copying interior values into ghosts:
#   p[1] = p[2], p[0] = p[1]   (left)
#   p[Nx+2] = p[Nx+1], p[Nx+3] = p[Nx+2]   (right)
##########################################
def apply_neumann_p(p, Nx):
    """
    Apply simple homogeneous Neumann-style BCs for p:
    copy interior boundary values into ghosts (p_x ‚âà 0).
    """
    # Left side
    p[1] = p[2]
    p[0] = p[1]

    # Right side
    p[Nx+2] = p[Nx+1]
    p[Nx+3] = p[Nx+2]

##########################################
# Main solver for 1D wave equation (first-order form)
# with *direct Neumann BCs on u*
##########################################
def solve_wave_1d_BW_char_neumann(
    c=1.0,
    L=1.0,
    Nx=200,
    Tfinal=1.0,
    CFL=0.5,
    hL=lambda t: 0.0,
    hR=lambda t: 0.0
):
    dx = L / Nx
    # time step from CFL: use |Œª_max| = c
    dt = CFL * dx / c
    Nt = int(np.ceil(Tfinal / dt))
    dt = Tfinal / Nt  # snap to hit Tfinal exactly

    # Total size with 2 ghost cells on each side
    Ntot = Nx + 4
    i_start, i_end = 2, Nx+1  # interior indices

    # Allocate fields with ghost cells
    u = np.zeros(Ntot)
    p = np.zeros(Ntot)
    r = np.zeros(Ntot)

    # Coordinates at cell centers (for plotting)
    # Interior cells i=2..Nx+1 cover [0,L]
    x = (np.arange(Ntot) - 2 + 0.5) * dx

    # Initial condition: e.g., a Gaussian bump for u, zero initial p
    x_phys = x[i_start:i_end+1]
    u[i_start:i_end+1] = np.exp(-100.0 * (x_phys - 0.5*L)**2)
    p[i_start:i_end+1] = 0.0

    # Apply Neumann BC on u and p at t=0
    apply_neumann_u(u, Nx, dx, 0.0, hL, hR)
    apply_neumann_p(p, Nx)

    # Initialize r = c u_x using centered difference (uses ghost cells of u)
    r[i_start:i_end+1] = c * (u[i_start+1:i_end+2] - u[i_start-1:i_end]) / (2.0*dx)

    # Time stepping
    for n in range(Nt):
        t = n*dt

        # 1) Enforce Neumann BCs on u and p for current time
        apply_neumann_u(u, Nx, dx, t,   hL, hR)
        apply_neumann_p(p, Nx)

        # 2) Recompute r = c u_x with current u (and its ghosts)
        r[i_start:i_end+1] = c * (u[i_start+1:i_end+2] - u[i_start-1:i_end]) / (2.0*dx)

        # 3) Transform (p, r) into characteristic variables:
        #    w_- = 0.5*(p + r), w_+ = 0.5*(p - r)
        w_minus = 0.5*(p + r)  # left-going, speed -c
        w_plus  = 0.5*(p - r)  # right-going, speed +c

        # 4) Beam‚ÄìWarming update on each characteristic field
        #    w_-: w_-_t + (-c) w_-_x = 0  (a = -c)
        #    w_+: w_+_t + (+c) w_+_x = 0  (a = +c)
        w_minus_new = BW_timestep_scalar(w_minus, a=-c, dx=dx, dt=dt,
                                         i_start=i_start, i_end=i_end)
        w_plus_new  = BW_timestep_scalar(w_plus,  a=+c, dx=dx, dt=dt,
                                         i_start=i_start, i_end=i_end)

        # 5) Transform back to p, r
        p = w_minus_new + w_plus_new
        r = w_minus_new - w_plus_new

        # 6) Update u with u_t = p (forward Euler in time)
        u[i_start:i_end+1] += dt * p[i_start:i_end+1]

        # 7) Re-enforce Neumann on u at t+dt (for the *next* step)
        apply_neumann_u(u, Nx, dx, t+dt, hL, hR)

    # Extract physical interior for plotting and output
    x_phys = x[i_start:i_end+1]
    u_phys = u[i_start:i_end+1]
    p_phys = p[i_start:i_end+1]
    r_phys = r[i_start:i_end+1]
    return x_phys, u_phys, p_phys, r_phys

##########################################
# Run a demo with homogeneous Neumann:
#   u_x(0,t) = u_x(L,t) = 0  (free ends)
##########################################
if __name__ == "__main__":
    c = 1.0
    L = 1.0
    Nx = 200
    Tfinal = 1.0
    CFL = 0.8

    # Homogeneous Neumann: hL = hR = 0 (free-end boundary)
    x, u, p, r = solve_wave_1d_BW_char_neumann(
        c=c, L=L, Nx=Nx, Tfinal=Tfinal, CFL=CFL,
        hL=lambda t: 0.0,
        hR=lambda t: 0.0
    )

    plt.figure(figsize=(6,4))
    plt.plot(x, u, label='u(x,T)')
    plt.xlabel('x'); plt.ylabel('u')
    plt.title('1D Wave Equation ‚Äì Beam‚ÄìWarming + Direct Neumann BCs')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

## Outflow (non-reflecting) boundary conditions

An **outflow** or **non-reflecting** boundary condition is meant to let waves exit the domain with minimal reflection.

For the 1D constant‚Äìcoefficient wave equation, the exact incoming characteristic at a boundary is:

- At $x=0$: incoming field is $w_+$.
- At $x=L$: incoming field is $w_-$.

A simple (and for this linear problem, exact) non-reflecting condition is to **set the incoming characteristic equal to zero** (or to a prescribed far-field value).

### Characteristic form

- At the **left boundary $x=0$**, a pure outflow to the left (no waves entering from the left) corresponds to
  $$
  \boxed{
  w_+(0,t) = 0
  }
  $$
  (or more generally $w_+(0,t)=w_{+,{\infty}}(t)$).

- At the **right boundary $x=L$**, a pure outflow to the right corresponds to
  $$
  \boxed{
  w_-(L,t) = 0
  }
  $$
  (or more generally $w_-(L,t)=w_{-,{\infty}}(t)$).

In terms of $u$, recall
$$
w_- = p - r = u_t - c\,u_x.
$$

So at the right boundary, the condition $w_-(L,t)=0$ is equivalent to the **Sommerfeld radiation condition**
$$
u_t(L,t) - c\,u_x(L,t) = 0,
$$
which says that the left‚Äìgoing component vanishes; only the right‚Äìgoing wave remains.

Similarly at the left boundary, $w_+(0,t)=0$ is
$$
u_t(0,t) + c\,u_x(0,t) = 0,
$$
eliminating the right‚Äìgoing component entering from the left.

```python
import numpy as np
import matplotlib.pyplot as plt

##########################################
# 1D wave equation in first-order form
#   u_tt = c^2 u_xx
# Introduce:
#   p = u_t
#   r = c u_x
#
# First-order system (for [p, r]):
#   p_t - c r_x = 0
#   r_t - c p_x = 0
#
# Characteristic variables:
#   w_- = 0.5*(p + r)   (left-going, speed -c)
#   w_+ = 0.5*(p - r)   (right-going, speed +c)
#
# Reconstruction:
#   p = w_- + w_+
#   r = w_- - w_+
#
# The code below first implements *direct Neumann BCs* on u (your original),
# and then adds a separate solver with *outflow (Sommerfeld) BCs*
# implemented in characteristic variables.
##########################################

##########################################
# Beam‚ÄìWarming scalar stencils (with ghost cells)
##########################################
def BW_derivatives(q, a, dx, i_start, i_end):
    """
    Compute first and second upwind derivatives for Beam‚ÄìWarming
    for scalar q on indices i_start...i_end (inclusive),
    assuming ghost cells outside that range are valid.
    """
    D1 = np.zeros_like(q)
    D2 = np.zeros_like(q)

    if a == 0.0:
        return D1, D2

    if a > 0:
        # Backward-biased (i, i-1, i-2)
        i = np.arange(i_start, i_end+1)
        D1[i] = (3.0*q[i] - 4.0*q[i-1] + q[i-2]) / (2.0*dx)
        D2[i] = (q[i] - 2.0*q[i-1] + q[i-2]) / (dx*dx)
    else:
        # Forward-biased (i, i+1, i+2)
        i = np.arange(i_start, i_end+1)
        D1[i] = (-3.0*q[i] + 4.0*q[i+1] - q[i+2]) / (2.0*dx)
        D2[i] = (q[i] - 2.0*q[i+1] + q[i+2]) / (dx*dx)

    return D1, D2

def BW_timestep_scalar(q, a, dx, dt, i_start, i_end):
    """
    One Beam‚ÄìWarming step for q_t + a q_x = 0 on interior cells.
    Uses the Taylor expansion:
       q^{n+1} = q^n - a dt D1 + 0.5 (a dt)^2 D2
    """
    D1, D2 = BW_derivatives(q, a, dx, i_start, i_end)
    q_new = q.copy()
    q_new[i_start:i_end+1] = (
        q[i_start:i_end+1]
        - a*dt*D1[i_start:i_end+1]
        + 0.5*(a*dt)*(a*dt)*D2[i_start:i_end+1]
    )
    return q_new

##########################################
# Direct Neumann BCs on u with ghost cells (your original)
##########################################
def apply_neumann_u(u, Nx, dx, t, hL_func, hR_func):
    """
    Apply Neumann BCs to u at left/right via ghost cells.

    hL_func, hR_func: functions of time t returning u_x(0,t), u_x(L,t)

    Ghost layout:
      indices: 0,1 | 2,...,Nx+1 | Nx+2,Nx+3
      physical interior: 2 ... Nx+1
    """
    hL = hL_func(t)
    hR = hR_func(t)

    # Left boundary: (u[2] - u[1]) / dx = hL
    u[1] = u[2] - dx * hL
    u[0] = 2.0 * u[1] - u[2]

    # Right boundary: (u[Nx+2] - u[Nx+1]) / dx = hR
    u[Nx+2] = u[Nx+1] + dx * hR
    u[Nx+3] = 2.0 * u[Nx+2] - u[Nx+1]

def apply_neumann_p(p, Nx):
    """
    Apply simple homogeneous Neumann-style BCs for p:
    copy interior boundary values into ghosts (p_x ‚âà 0).
    """
    # Left side
    p[1] = p[2]
    p[0] = p[1]

    # Right side
    p[Nx+2] = p[Nx+1]
    p[Nx+3] = p[Nx+2]

##########################################
# Original Neumann-based solver
##########################################
def solve_wave_1d_BW_char_neumann(
    c=1.0,
    L=1.0,
    Nx=200,
    Tfinal=1.0,
    CFL=0.5,
    hL=lambda t: 0.0,
    hR=lambda t: 0.0
):
    dx = L / Nx
    dt = CFL * dx / c
    Nt = int(np.ceil(Tfinal / dt))
    dt = Tfinal / Nt

    Ntot = Nx + 4
    i_start, i_end = 2, Nx+1

    u = np.zeros(Ntot)
    p = np.zeros(Ntot)
    r = np.zeros(Ntot)

    x = (np.arange(Ntot) - 2 + 0.5) * dx

    x_phys = x[i_start:i_end+1]
    u[i_start:i_end+1] = np.exp(-100.0 * (x_phys - 0.5*L)**2)
    p[i_start:i_end+1] = 0.0

    apply_neumann_u(u, Nx, dx, 0.0, hL, hR)
    apply_neumann_p(p, Nx)

    r[i_start:i_end+1] = c * (u[i_start+1:i_end+2] - u[i_start-1:i_end]) / (2.0*dx)

    for n in range(Nt):
        t = n*dt

        apply_neumann_u(u, Nx, dx, t,   hL, hR)
        apply_neumann_p(p, Nx)

        r[i_start:i_end+1] = c * (u[i_start+1:i_end+2] - u[i_start-1:i_end]) / (2.0*dx)

        w_minus = 0.5*(p + r)
        w_plus  = 0.5*(p - r)

        w_minus_new = BW_timestep_scalar(w_minus, a=-c, dx=dx, dt=dt,
                                         i_start=i_start, i_end=i_end)
        w_plus_new  = BW_timestep_scalar(w_plus,  a=+c, dx=dx, dt=dt,
                                         i_start=i_start, i_end=i_end)

        p = w_minus_new + w_plus_new
        r = w_minus_new - w_plus_new

        u[i_start:i_end+1] += dt * p[i_start:i_end+1]

        apply_neumann_u(u, Nx, dx, t+dt, hL, hR)

    x_phys = x[i_start:i_end+1]
    u_phys = u[i_start:i_end+1]
    p_phys = p[i_start:i_end+1]
    r_phys = r[i_start:i_end+1]
    return x_phys, u_phys, p_phys, r_phys

##########################################
# NEW: Outflow (non-reflecting) boundary conditions
#
# Sommerfeld form:
#   at x=0: u_t + c u_x = 0  => w_+(0,t) = 0
#   at x=L: u_t - c u_x = 0  => w_-(L,t) = 0
#
# We implement this directly in *characteristic variables*:
#   - At left boundary, incoming w_+ set to gL(t) (usually 0),
#     outgoing w_- copied from interior.
#   - At right boundary, incoming w_- set to gR(t) (usually 0),
#     outgoing w_+ copied from interior.
##########################################
def apply_outflow_char(w_minus, w_plus, Nx, t, gL_func, gR_func):
    """
    Apply outflow (Sommerfeld) BCs to characteristic variables.

    Ghost layout:
      indices: 0,1 | 2,...,Nx+1 | Nx+2,Nx+3
      physical interior: 2 ... Nx+1

    At left boundary x=0 (incoming char is w_+):
      w_+[1]   = gL(t)
      w_-[1]   = w_-[2]      (outgoing from interior)
      w_+[0]   = w_+[1]
      w_-[0]   = w_-[1]

    At right boundary x=L (incoming char is w_-):
      w_-[Nx+2] = gR(t)
      w_+[Nx+2] = w_+[Nx+1]  (outgoing from interior)
      w_-[Nx+3] = w_-[Nx+2]
      w_+[Nx+3] = w_+[Nx+2]
    """
    gL = gL_func(t)
    gR = gR_func(t)

    # Left boundary
    w_plus[1]  = gL         # incoming
    w_minus[1] = w_minus[2] # outgoing from interior

    w_plus[0]  = w_plus[1]
    w_minus[0] = w_minus[1]

    # Right boundary
    w_minus[Nx+2] = gR           # incoming
    w_plus[Nx+2]  = w_plus[Nx+1] # outgoing from interior

    w_minus[Nx+3] = w_minus[Nx+2]
    w_plus[Nx+3]  = w_plus[Nx+2]

##########################################
# Solver with outflow (Sommerfeld) BCs
##########################################
def solve_wave_1d_BW_char_outflow(
    c=1.0,
    L=1.0,
    Nx=200,
    Tfinal=1.0,
    CFL=0.5,
    gL=lambda t: 0.0,  # incoming w_+ at x=0
    gR=lambda t: 0.0   # incoming w_- at x=L
):
    dx = L / Nx
    dt = CFL * dx / c
    Nt = int(np.ceil(Tfinal / dt))
    dt = Tfinal / Nt

    Ntot = Nx + 4
    i_start, i_end = 2, Nx+1

    u = np.zeros(Ntot)
    p = np.zeros(Ntot)
    r = np.zeros(Ntot)

    x = (np.arange(Ntot) - 2 + 0.5) * dx

    # Initial condition: Gaussian in u, p=0, r from u_x
    x_phys = x[i_start:i_end+1]
    u[i_start:i_end+1] = np.exp(-100.0 * (x_phys - 0.5*L)**2)
    p[i_start:i_end+1] = 0.0
    # For initial r, approximate c u_x using interior neighbors.
    # We don't strictly need ghost u for this if we avoid r at the first/last
    # cell in the derivative, but we'll just do a one-sided at ends:
    r[i_start:i_end+1] = c * (u[i_start+1:i_end+2] - u[i_start-1:i_end]) / (2.0*dx)

    for n in range(Nt):
        t = n*dt

        # 1) Build characteristic variables
        w_minus = 0.5*(p + r)
        w_plus  = 0.5*(p - r)

        # 2) Apply outflow BCs to w_- and w_+
        apply_outflow_char(w_minus, w_plus, Nx, t, gL, gR)

        # 3) Beam‚ÄìWarming update for w_- and w_+
        w_minus_new = BW_timestep_scalar(w_minus, a=-c, dx=dx, dt=dt,
                                         i_start=i_start, i_end=i_end)
        w_plus_new  = BW_timestep_scalar(w_plus,  a=+c, dx=dx, dt=dt,
                                         i_start=i_start, i_end=i_end)

        # 4) Transform back to p, r
        p = w_minus_new + w_plus_new
        r = w_minus_new - w_plus_new

        # 5) Update u with u_t = p (forward Euler)
        u[i_start:i_end+1] += dt * p[i_start:i_end+1]

        # (No direct BC on u or p; boundaries handled via characteristics.)

    x_phys = x[i_start:i_end+1]
    u_phys = u[i_start:i_end+1]
    p_phys = p[i_start:i_end+1]
    r_phys = r[i_start:i_end+1]
    return x_phys, u_phys, p_phys, r_phys

##########################################
# Demo: outflow BCs (no incoming waves)
##########################################
if __name__ == "__main__":
    c = 1.0
    L = 1.0
    Nx = 200
    Tfinal = 0.5
    CFL = 0.8

    # Pure outflow: incoming characteristics set to 0
    x, u, p, r = solve_wave_1d_BW_char_outflow(
        c=c, L=L, Nx=Nx, Tfinal=Tfinal, CFL=CFL,
        gL=lambda t: 0.0,
        gR=lambda t: 0.0
    )

    plt.figure(figsize=(6,4))
    plt.plot(x, u, label='u(x,T)')
    plt.xlabel('x'); plt.ylabel('u')
    plt.title('1D Wave Equation ‚Äì Beam‚ÄìWarming + Outflow (Sommerfeld) BCs')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

## Summary (practical use)

For the 1-D wave equation in first-order form, with
$$
p = u_t,\quad r = c\,u_x,\quad
w_+ = p + r,\quad w_- = p - r,
$$
we have:

- **Dirichlet displacement BCs (direct form)**:
  $$
  u(0,t)=g_0(t), \quad u(L,t)=g_L(t),
  $$
  implying
  $$
  p(0,t)=g_0'(t),\quad p(L,t)=g_L'(t),
  $$
  with $u_x(0,t)$ and $u_x(L,t)$ enforced via the spatial stencil.

- **Dirichlet in characteristic form**:
  - at $x=0$: specify incoming $w_+ = p+r$,
  - at $x=L$: specify incoming $w_- = p-r$.

- **Neumann (direct form)**:
  $$
  u_x(0,t) = h_0(t), \qquad u_x(L,t) = h_L(t),
  $$
  i.e.
  $$
  r(0,t) = c h_0(t), \qquad r(L,t) = c h_L(t).
  $$

- **Neumann in characteristic form**:
  $$
  w_+(0,t) = w_-(0,t) + 2\,c h_0(t), \qquad
  w_-(L,t) = w_+(L,t) - 2\,c h_L(t).
  $$

- **Periodic BCs (direct form)**:
  $$
  u(0,t)=u(L,t),\quad u_x(0,t)=u_x(L,t),
  $$
  equivalently
  $$
  p(0,t)=p(L,t),\quad r(0,t)=r(L,t).
  $$

- **Periodic in characteristic form**:
  $$
  w_\pm(0,t)=w_\pm(L,t).
  $$

- **Outflow / non-reflecting in characteristic form**:
  $$
  w_+(0,t) = 0 \quad\text{(no incoming from the left)},\qquad
  w_-(L,t) = 0 \quad\text{(no incoming from the right)}.
  $$

This is the form a numerical solver actually uses; the physical (direct) BCs on $u$ (Dirichlet/Neumann/periodic/outflow) are translated into conditions on $w_+$ and $w_-$ at the boundaries.

# 2D Wave Equation with Direct Boundary Conditions

We consider the 2D wave equation on a rectangular domain
$$
(x,y) \in [0,L_x] \times [0,L_y],
$$
with constant wave speed $c$:
$$
u_{tt} = c^2 \left( u_{xx} + u_{yy} \right).
$$

---

## First‚Äìorder form in 2D

Introduce first‚Äìorder variables:
$$
p = u_t, 
\qquad
r = c\,u_x,
\qquad
s = c\,u_y.
$$

Then the 2D wave equation can be written as the first‚Äìorder system
$$
\begin{aligned}
u_t &= p, \\[2mm]
p_t - c\,r_x - c\,s_y &= 0, \\[2mm]
r_t - c\,p_x &= 0, \\[2mm]
s_t - c\,p_y &= 0.
\end{aligned}
$$

Here:
- $u(x,y,t)$ is the displacement,
- $p(x,y,t)$ is the velocity,
- $r(x,y,t)$ and $s(x,y,t)$ are (scaled) flux variables in the $x$ and $y$ directions, respectively.

The relations
$$
p = u_t, \qquad r = c\,u_x, \qquad s = c\,u_y
$$
allow us to recover $u_x$ and $u_y$ once $r$ and $s$ are known.

---

## Direct (physical) boundary conditions in 2D

Boundary conditions are imposed on the **boundary of the rectangle**:
- left: $x=0$,
- right: $x=L_x$,
- bottom: $y=0$,
- top: $y=L_y$.

These conditions are most naturally written in terms of $u$ and its normal derivative $u_n$.

### 1. Dirichlet (displacement) boundary conditions

Specify the displacement $u$ on each side:
$$
\begin{aligned}
u(0,y,t)   &= g_{\text{left}}(y,t),   & 0 \le y \le L_y, \\
u(L_x,y,t) &= g_{\text{right}}(y,t),  & 0 \le y \le L_y, \\
u(x,0,t)   &= g_{\text{bottom}}(x,t), & 0 \le x \le L_x, \\
u(x,L_y,t) &= g_{\text{top}}(x,t),    & 0 \le x \le L_x.
\end{aligned}
$$

These conditions fix the **value of $u$** on the boundary.

In terms of first‚Äìorder variables:
- $p = u_t$ on the boundary is given by the **time derivative** of the boundary data:
  $$
  p(0,y,t) = \frac{\partial g_{\text{left}}}{\partial t}(y,t),
  \quad \text{etc.}
  $$
- $r = c\,u_x$ and $s = c\,u_y$ are obtained from the **spatial derivatives** of $u$ near the boundary (computed numerically using ghost cells / one‚Äìsided stencils).

---

### 2. Neumann (normal derivative / flux) boundary conditions

Neumann BCs specify the **normal derivative of $u$** on each side:

- On $x = 0$ (left): the outward normal is $(-1,0)$, so
  $$
  u_x(0,y,t) = h_{\text{left}}(y,t).
  $$
- On $x = L_x$ (right): normal $(+1,0)$,
  $$
  u_x(L_x,y,t) = h_{\text{right}}(y,t).
  $$
- On $y = 0$ (bottom): normal $(0,-1)$,
  $$
  u_y(x,0,t) = h_{\text{bottom}}(x,t).
  $$
- On $y = L_y$ (top): normal $(0,+1)$,
  $$
  u_y(x,L_y,t) = h_{\text{top}}(x,t).
  $$

In terms of $(p,r,s)$:
$$
r(0,y,t) = c\,h_{\text{left}}(y,t),
\qquad
r(L_x,y,t) = c\,h_{\text{right}}(y,t),
$$
$$
s(x,0,t) = c\,h_{\text{bottom}}(x,t),
\qquad
s(x,L_y,t) = c\,h_{\text{top}}(x,t).
$$

The displacement $u$ itself on the boundary is then determined up to an integration in space (and must be made consistent with the initial condition).

---

### 3. Periodic boundary conditions

On a 2D periodic domain, opposite sides are identified:

- Periodic in $x$:
  $$
  u(0,y,t) = u(L_x,y,t), \qquad u_x(0,y,t) = u_x(L_x,y,t),
  $$
  and similarly for $p$, $r$, $s$.

- Periodic in $y$:
  $$
  u(x,0,t) = u(x,L_y,t), \qquad u_y(x,0,t) = u_y(x,L_y,t),
  $$
  and similarly for $p$, $r$, $s$.

In a fully periodic box, both pairs of opposite sides use these identifications.

---

### 4. Outflow (non‚Äìreflecting / Sommerfeld‚Äìtype) conditions

A common approximate non‚Äìreflecting condition in 2D is applied in the **normal direction** to each side. For example, on the right boundary $x=L_x$:
$$
u_t(L_x,y,t) - c\,u_x(L_x,y,t) = g_{\text{out,right}}(y,t),
$$
which, for $g_{\text{out,right}} = 0$, suppresses incoming waves from the right and lets waves travel outward.

Similarly, on the left boundary $x=0$:
$$
u_t(0,y,t) + c\,u_x(0,y,t) = g_{\text{out,left}}(y,t),
$$
and analogously in the $y$-direction on the top and bottom boundaries:
$$
u_t(x,0,t) + c\,u_y(x,0,t) = g_{\text{out,bottom}}(x,t),
$$
$$
u_t(x,L_y,t) - c\,u_y(x,L_y,t) = g_{\text{out,top}}(x,t).
$$

In terms of first‚Äìorder variables:
- On $x=L_x$:
  $$
  p(L_x,y,t) - r(L_x,y,t) = g_{\text{out,right}}(y,t),
  $$
- On $x=0$:
  $$
  p(0,y,t) + r(0,y,t) = g_{\text{out,left}}(y,t),
  $$
and analogously replacing $r$ with $s$ for boundaries in the $y$ direction.

---

## Summary

- The **2D wave equation**
  $$
  u_{tt} = c^2(u_{xx} + u_{yy})
  $$
  can be written as a **first‚Äìorder hyperbolic system** in $(u,p,r,s)$.

- **Direct boundary conditions** are naturally expressed in terms of $u$ and its normal derivative $u_n$:
  - Dirichlet: specify $u$ on each side.
  - Neumann: specify the normal derivative $u_n$ (and hence $r$ or $s$).
  - Periodic: identify opposite sides.
  - Outflow: impose Sommerfeld‚Äìtype conditions relating $u_t$ and $u_n$ to suppress reflections.

These direct BCs on $u$ and $u_x,u_y$ induce corresponding conditions on $p$, $r$, and $s$ in the first‚Äìorder formulation used for numerical schemes.


```python
import numpy as np
import matplotlib.pyplot as plt

##########################################
# 1D Beam‚ÄìWarming helpers (unchanged)
##########################################
def BW_derivatives(q, a, dx, i_start, i_end):
    """
    Compute first and second upwind derivatives for Beam‚ÄìWarming
    for scalar q on indices i_start...i_end (inclusive),
    assuming ghost cells outside that range are valid.

    q : 1D array with ghost cells
    a : scalar advection speed
    dx: grid spacing
    i_start, i_end: integer indices of interior cells (e.g., 2, Nx+1)

    Returns:
        D1, D2 (arrays same shape as q, values set only on interior indices)
    """
    D1 = np.zeros_like(q)
    D2 = np.zeros_like(q)

    if a == 0.0:
        return D1, D2

    if a > 0:
        # Backward-biased (i, i-1, i-2)
        i = np.arange(i_start, i_end+1)
        D1[i] = (3.0*q[i] - 4.0*q[i-1] + q[i-2]) / (2.0*dx)
        D2[i] = (q[i] - 2.0*q[i-1] + q[i-2]) / (dx*dx)
    else:
        # Forward-biased (i, i+1, i+2)
        i = np.arange(i_start, i_end+1)
        D1[i] = (-3.0*q[i] + 4.0*q[i+1] - q[i+2]) / (2.0*dx)
        D2[i] = (q[i] - 2.0*q[i+1] + q[i+2]) / (dx*dx)

    return D1, D2

def BW_timestep_scalar(q, a, dx, dt, i_start, i_end):
    """
    One Beam‚ÄìWarming step for q_t + a q_x = 0 on interior cells.
    Uses the Taylor expansion:
       q^{n+1} = q^n - a dt D1 + 0.5 (a dt)^2 D2
    """
    D1, D2 = BW_derivatives(q, a, dx, i_start, i_end)
    q_new = q.copy()
    q_new[i_start:i_end+1] = (
        q[i_start:i_end+1]
        - a*dt*D1[i_start:i_end+1]
        + 0.5*(a*dt)*(a*dt)*D2[i_start:i_end+1]
    )
    return q_new

##########################################
# 2D Dirichlet BCs for u and p
#
# Layout in x and y (with ghost cells):
#   x-indices: 0,1 | 2,...,Nx+1 | Nx+2,Nx+3  (Nx interior)
#   y-indices: 0,1 | 2,...,Ny+1 | Ny+2,Ny+3  (Ny interior)
#
# Homogeneous Dirichlet:
#   u = 0 on all four sides.
# For p = u_t, we also set p = 0 at boundary, which corresponds
# to the boundary value not changing in time.
##########################################
def apply_dirichlet_u_2d(u, Nx, Ny):
    """
    Homogeneous Dirichlet u = 0 on all boundaries via ghost + boundary cells.

    u: 2D array with shape (Nx+4, Ny+4)
    Interior cells: i=2..Nx+1, j=2..Ny+1
    """
    # Left and right boundaries (including ghosts)
    u[0:2, :]        = 0.0      # left ghosts + left boundary
    u[Nx+2:Nx+4, :]  = 0.0      # right ghosts + right boundary

    # Bottom and top boundaries (including ghosts)
    u[:, 0:2]        = 0.0      # bottom ghosts + bottom boundary
    u[:, Ny+2:Ny+4]  = 0.0      # top ghosts + top boundary


def apply_dirichlet_p_2d(p, Nx, Ny):
    """
    Homogeneous Dirichlet for p = u_t on all boundaries (p = 0 at boundary),
    with *second-order* ghost-cell extrapolation in the normal direction.

    Layout:
      x-indices: 0,1 | 2,...,Nx+1 | Nx+2,Nx+3
      y-indices: 0,1 | 2,...,Ny+1 | Ny+2,Ny+3

    Interior: i = 2..Nx+1, j = 2..Ny+1
    Boundary lines: i=2 (left), i=Nx+1 (right), j=2 (bottom), j=Ny+1 (top)
    """

    # --------------------------------------------------
    # 1) Enforce p = 0 on the *interior boundary lines*
    # --------------------------------------------------
    # Left/right vertical boundaries
    p[2,  :]      = 0.0
    p[Nx+1, :]    = 0.0

    # Bottom/top horizontal boundaries
    p[:,  2]      = 0.0
    p[:,  Ny+1]   = 0.0

    # --------------------------------------------------
    # 2) Second-order ghost cells in x (left/right)
    #    For each fixed j, apply 1D quadratic extrapolation in x.
    #    Using p0 = 0 at boundary, and interior p1, p2:
    #       p_ghost1 = 3 p0 - 3 p1 + p2 = -3 p1 + p2
    #       p_ghost2 = 6 p0 - 8 p1 + 3 p2 = -8 p1 + 3 p2
    # --------------------------------------------------
    for j in range(0, Ny+4):
        # Left side (i direction): interior indices i=2,3,4
        p0L = p[2, j]           # 0, by construction
        p1L = p[3, j]
        p2L = p[4, j]

        p[1, j] = 3.0*p0L - 3.0*p1L + p2L    # first ghost (i=1)
        p[0, j] = 6.0*p0L - 8.0*p1L + 3.0*p2L  # second ghost (i=0)

        # Right side: interior indices i=Nx+1, Nx, Nx-1
        p0R = p[Nx+1, j]       # 0, by construction
        p1R = p[Nx,   j]
        p2R = p[Nx-1, j]

        p[Nx+2, j] = 3.0*p0R - 3.0*p1R + p2R      # first ghost (i=Nx+2)
        p[Nx+3, j] = 6.0*p0R - 8.0*p1R + 3.0*p2R  # second ghost (i=Nx+3)

    # --------------------------------------------------
    # 3) Second-order ghost cells in y (bottom/top)
    #    For each fixed i, apply 1D quadratic extrapolation in y.
    #    Same formulas, now in j.
    # --------------------------------------------------
    for i in range(0, Nx+4):
        # Bottom side: j=2,3,4
        p0B = p[i, 2]           # 0
        p1B = p[i, 3]
        p2B = p[i, 4]

        p[i, 1] = 3.0*p0B - 3.0*p1B + p2B       # first ghost (j=1)
        p[i, 0] = 6.0*p0B - 8.0*p1B + 3.0*p2B   # second ghost (j=0)

        # Top side: j=Ny+1, Ny, Ny-1
        p0T = p[i, Ny+1]        # 0
        p1T = p[i, Ny]
        p2T = p[i, Ny-1]

        p[i, Ny+2] = 3.0*p0T - 3.0*p1T + p2T      # first ghost (j=Ny+2)
        p[i, Ny+3] = 6.0*p0T - 8.0*p1T + 3.0*p2T  # second ghost (j=Ny+3)


##########################################
# 2D Wave solver using dimensional splitting
# and 1D Beam‚ÄìWarming in x and y.
##########################################
def solve_wave_2d_BW_dirichlet(
    c=1.0,
    Lx=1.0,
    Ly=1.0,
    Nx=100,
    Ny=100,
    Tfinal=1.0,
    CFL=0.3
):
    """
    2D wave equation:
        u_tt = c^2 (u_xx + u_yy)
    in first-order form with variables:
        u, p = u_t, r = c u_x, s = c u_y

    Using dimensional splitting + 1D Beam‚ÄìWarming in x and y, with
    homogeneous Dirichlet BCs u = 0 on all sides of the box.
    """
    dx = Lx / Nx
    dy = Ly / Ny
    # conservative CFL choice for split scheme
    dt = CFL * min(dx, dy) / c
    Nt = int(np.ceil(Tfinal / dt))
    dt = Tfinal / Nt

    # Indices and sizes with 2 ghost cells each side
    Nxtot = Nx + 4
    Nytot = Ny + 4
    i_start, i_end = 2, Nx+1  # interior x
    j_start, j_end = 2, Ny+1  # interior y

    # Allocate arrays
    u = np.zeros((Nxtot, Nytot))
    p = np.zeros((Nxtot, Nytot))
    r = np.zeros((Nxtot, Nytot))
    s = np.zeros((Nxtot, Nytot))

    # Coordinates for interior cells (for ICs & plotting)
    x = (np.arange(Nxtot) - 2 + 0.5) * dx
    y = (np.arange(Nytot) - 2 + 0.5) * dy
    X, Y = np.meshgrid(x, y, indexing='ij')

    # Initial condition: 2D Gaussian bump in the center, zero velocity
    X_phys = X[i_start:i_end+1, j_start:j_end+1]
    Y_phys = Y[i_start:i_end+1, j_start:j_end+1]
    u[i_start:i_end+1, j_start:j_end+1] = np.exp(
        -100.0 * ((X_phys - 0.5*Lx)**2 + (Y_phys - 0.5*Ly)**2)
    )
    p[i_start:i_end+1, j_start:j_end+1] = 0.0

    # Apply Dirichlet BCs initially
    apply_dirichlet_u_2d(u, Nx, Ny)
    apply_dirichlet_p_2d(p, Nx, Ny)

    # Initialize r = c u_x, s = c u_y using centered differences
    r[i_start:i_end+1, j_start:j_end+1] = c * (
        u[i_start+1:i_end+2, j_start:j_end+1] - u[i_start-1:i_end, j_start:j_end+1]
    ) / (2.0*dx)

    s[i_start:i_end+1, j_start:j_end+1] = c * (
        u[i_start:i_end+1, j_start+1:j_end+2] - u[i_start:i_end+1, j_start-1:j_end]
    ) / (2.0*dy)

    # Time-stepping
    for n in range(Nt):
        t = n*dt

        # Enforce Dirichlet BCs on u, p at the start of step
        apply_dirichlet_u_2d(u, Nx, Ny)
        apply_dirichlet_p_2d(p, Nx, Ny)

        # Recompute r and s from u
        r[i_start:i_end+1, j_start:j_end+1] = c * (
            u[i_start+1:i_end+2, j_start:j_end+1] - u[i_start-1:i_end, j_start:j_end+1]
        ) / (2.0*dx)
        s[i_start:i_end+1, j_start:j_end+1] = c * (
            u[i_start:i_end+1, j_start+1:j_end+2] - u[i_start:i_end+1, j_start-1:j_end]
        ) / (2.0*dy)

        # ============================
        # X-SWEEP: update (p,r) along x
        # ============================
        p_x_sweep = p.copy()
        r_x_sweep = r.copy()

        for j in range(j_start, j_end+1):  # loop over y rows
            # 1D characteristic variables in x: w_-^x, w_+^x
            px = p[:, j]
            rx = r[:, j]

            w_minus_x = 0.5 * (px + rx)  # left-going, speed -c along x
            w_plus_x  = 0.5 * (px - rx)  # right-going, speed +c along x

            # Beam‚ÄìWarming in x for w_- and w_+
            w_minus_x_new = BW_timestep_scalar(w_minus_x, a=-c, dx=dx, dt=dt,
                                               i_start=i_start, i_end=i_end)
            w_plus_x_new  = BW_timestep_scalar(w_plus_x,  a=+c, dx=dx, dt=dt,
                                               i_start=i_start, i_end=i_end)

            # Reconstruct p,r for this row
            p_x_sweep[:, j] = w_minus_x_new + w_plus_x_new
            r_x_sweep[:, j] = w_minus_x_new - w_plus_x_new

        # ============================
        # Y-SWEEP: update (p,s) along y
        # ============================
        p_new = p_x_sweep.copy()
        r_new = r_x_sweep.copy()   # r doesn't change in y-sweep (split)
        s_new = s.copy()

        for i in range(i_start, i_end+1):  # loop over x columns
            py = p_x_sweep[i, :]
            sy = s[i, :]

            w_minus_y = 0.5 * (py + sy)  # "down-going", speed -c along y
            w_plus_y  = 0.5 * (py - sy)  # "up-going",   speed +c along y

            # Beam‚ÄìWarming in y for w_- and w_+
            w_minus_y_new = BW_timestep_scalar(w_minus_y, a=-c, dx=dy, dt=dt,
                                               i_start=j_start, i_end=j_end)
            w_plus_y_new  = BW_timestep_scalar(w_plus_y,  a=+c, dx=dy, dt=dt,
                                               i_start=j_start, i_end=j_end)

            # Reconstruct p,s for this column
            p_new[i, :] = w_minus_y_new + w_plus_y_new
            s_new[i, :] = w_minus_y_new - w_plus_y_new

        # Update p,r,s
        p = p_new
        r = r_new
        s = s_new

        # Finally update u with u_t = p
        u[i_start:i_end+1, j_start:j_end+1] += dt * p[i_start:i_end+1, j_start:j_end+1]

    # Return physical interior and coordinates for plotting
    X_phys = X[i_start:i_end+1, j_start:j_end+1]
    Y_phys = Y[i_start:i_end+1, j_start:j_end+1]
    u_phys = u[i_start:i_end+1, j_start:j_end+1]

    return X_phys, Y_phys, u_phys

##########################################
# Demo
##########################################
if __name__ == "__main__":
    c = 1.0
    Lx, Ly = 1.0, 1.0
    Nx, Ny = 120, 120
    Tfinal = 0.7
    CFL = 0.10

    X, Y, U = solve_wave_2d_BW_dirichlet(
        c=c, Lx=Lx, Ly=Ly, Nx=Nx, Ny=Ny,
        Tfinal=Tfinal, CFL=CFL
    )

    plt.figure(figsize=(6,5))
    plt.pcolormesh(X, Y, U, shading="auto")
    plt.colorbar(label="u(x,y,T)")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("2D Wave Equation ‚Äì Beam‚ÄìWarming (split) + Dirichlet BCs")
    plt.tight_layout()
    plt.show()
```

# 2D Wave Equation with an Upper Boundary Moving to the Right

We consider the 2D wave equation on a rectangular domain
$$
(x,y) \in [0,L_x] \times [0,L_y],
$$
with constant wave speed $c$:
$$
u_{tt} = c^2\,(u_{xx} + u_{yy}).
$$

The field $u(x,y,t)$ is a scalar displacement. For example, it could represent the vertical displacement of a membrane or the horizontal displacement of a string-like structure in 2D.

---

## First‚Äìorder form (2D)

Introduce first‚Äìorder variables
$$
p = u_t, 
\qquad
r = c\,u_x,
\qquad
s = c\,u_y.
$$

Then the wave equation becomes the first‚Äìorder system
$$
\begin{aligned}
u_t &= p, \\[2mm]
p_t - c\,r_x - c\,s_y &= 0, \\[2mm]
r_t - c\,p_x &= 0, \\[2mm]
s_t - c\,p_y &= 0.
\end{aligned}
$$

The relations
$$
p = u_t, \quad r = c\,u_x, \quad s = c\,u_y
$$
allow us to reconstruct $u_x$ and $u_y$ from $(r,s)$.

---

## Domain and boundaries

We work on the rectangular box:
- left boundary: $x = 0$,
- right boundary: $x = L_x$,
- bottom boundary: $y = 0$,
- **top boundary: $y = L_y$ (this one will move to the right)**.

We will impose:
- homogeneous Dirichlet on left, right, and bottom,
- a **moving Dirichlet boundary condition** on the top.

---

## Homogeneous Dirichlet on the fixed boundaries

On three sides we take $u=0$ for simplicity:
$$
\begin{aligned}
u(0,y,t) &= 0, && 0 \le y \le L_y, \\
u(L_x,y,t) &= 0, && 0 \le y \le L_y, \\
u(x,0,t) &= 0, && 0 \le x \le L_x.
\end{aligned}
$$

In terms of the first‚Äìorder variables:
- $p = u_t = 0$ on these boundaries (the enforced value doesn‚Äôt change in time),
- $r$ and $s$ are determined from the spatial stencil near the boundary and the imposed $u$.

---

## Upper boundary moving to the right

To model an upper boundary whose **displacement pattern moves to the right** along $y=L_y$, we prescribe
$$
u(x,L_y,t) = g(x - V t),
$$
where:
- $V > 0$ is the **speed at which the boundary pattern moves to the right**,
- $g(\cdot)$ is some profile (e.g. a bump or wave shape).

This is a **Dirichlet boundary condition** along the top edge.

### Interpretation

- At $t=0$, $u(x,L_y,0) = g(x)$ sets the initial shape on the top boundary.
- At later times, the same shape has shifted by distance $V t$ to the right:
  $u(x,L_y,t) = g(x - V t)$.
- If $g$ is localized or periodic, this represents the upper boundary "sliding" or "waving" to the right and injecting energy into the interior.

---

## First‚Äìorder variables on the moving upper boundary

From the boundary data
$$
u_{\text{top}}(x,t) = g(x - V t),
$$
we can compute its **time derivative** (boundary velocity):
$$
p(x,L_y,t) = u_t(x,L_y,t) = \frac{\partial}{\partial t} g(x - V t)
= -V\,g'(x - V t).
$$

The normal derivative at the top is in the $+y$ direction. If we only prescribe $u$ (Dirichlet), then $s = c u_y$ is determined numerically from the interior stencil. The tangential derivative along $x$ is:
$$
u_x(x,L_y,t) = g'(x - V t),
$$
so at the top edge
$$
r(x,L_y,t) = c\,u_x(x,L_y,t) = c\,g'(x - V t).
$$

In numerical practice:

- We **impose** $u(x,L_y,t)=g(x-Vt)$ directly on the top boundary nodes.
- We set
  $$
  p(x,L_y,t) = -V\,g'(x - V t)
  $$
  consistently, either analytically (if $g$ is known) or via a finite-difference approximation in time.
- The normal derivative $u_y$ (and hence $s=c\,u_y$) is obtained from the interior values of $u$ using ghost cells or one‚Äìsided differences.

---

## Summary of the problem with a moving top boundary

We solve
$$
u_{tt} = c^2 (u_{xx} + u_{yy}), \quad (x,y) \in (0,L_x)\times(0,L_y), \quad t>0,
$$
with, for example,
$$
u(x,y,0) = u_0(x,y), \qquad u_t(x,y,0) = p_0(x,y),
$$
and boundary conditions:

- **Left** ($x=0$):
  $$
  u(0,y,t) = 0.
  $$

- **Right** ($x=L_x$):
  $$ 
  u(L_x,y,t) = 0.
  $$

- **Bottom** ($y=0$):
  $$
  u(x,0,t) = 0.
  $$

- **Top** ($y=L_y$), moving to the right:
  $$
  u(x,L_y,t) = g(x - V t).
  $$

In first‚Äìorder form, we track
$$
u, \quad p = u_t, \quad r = c u_x, \quad s = c u_y,
$$
and enforce the moving Dirichlet condition on $u$ at $y=L_y$, together with a consistent boundary value for $p = u_t$, while $r$ and $s$ at the boundary are obtained from spatial derivatives of $u$ using the chosen finite‚Äìdifference stencils and ghost cells.


```python
import numpy as np
import matplotlib.pyplot as plt

##########################################
# 1D Beam‚ÄìWarming helpers (unchanged)
##########################################
def BW_derivatives(q, a, dx, i_start, i_end):
    """
    Compute first and second upwind derivatives for Beam‚ÄìWarming
    for scalar q on indices i_start...i_end (inclusive),
    assuming ghost cells outside that range are valid.

    q      : 1D array with ghost cells
    a      : scalar advection speed
    dx     : grid spacing
    i_start, i_end : interior indices where we compute derivatives

    Returns:
        D1, D2 : first and second derivative approximations
    """
    D1 = np.zeros_like(q)
    D2 = np.zeros_like(q)

    if a == 0.0:
        return D1, D2

    if a > 0:
        # Backward-biased three-point stencil: (i, i-1, i-2)
        i = np.arange(i_start, i_end+1)
        D1[i] = (3.0*q[i] - 4.0*q[i-1] + q[i-2]) / (2.0*dx)
        D2[i] = (q[i] - 2.0*q[i-1] + q[i-2]) / (dx*dx)
    else:
        # Forward-biased three-point stencil: (i, i+1, i+2)
        i = np.arange(i_start, i_end+1)
        D1[i] = (-3.0*q[i] + 4.0*q[i+1] - q[i+2]) / (2.0*dx)
        D2[i] = (q[i] - 2.0*q[i+1] + q[i+2]) / (dx*dx)

    return D1, D2

def BW_timestep_scalar(q, a, dx, dt, i_start, i_end):
    """
    One Beam‚ÄìWarming step for the scalar advection equation:
        q_t + a q_x = 0
    on interior cells i_start...i_end.

    Time update (Taylor expansion using D1 and D2):
        q^{n+1} = q^n - a dt D1 + 0.5 (a dt)^2 D2
    """
    D1, D2 = BW_derivatives(q, a, dx, i_start, i_end)
    q_new = q.copy()
    q_new[i_start:i_end+1] = (
        q[i_start:i_end+1]
        - a*dt*D1[i_start:i_end+1]
        + 0.5*(a*dt)*(a*dt)*D2[i_start:i_end+1]
    )
    return q_new


##########################################
# 2D Dirichlet BCs for u and p with moving top boundary
#
# Grid layout in x and y (with 2 ghost cells per side):
#   x-indices: 0,1 | 2,...,Nx+1 | Nx+2,Nx+3  (Nx interior)
#   y-indices: 0,1 | 2,...,Ny+1 | Ny+2,Ny+3  (Ny interior)
#
# Physical interior:
#   i = 2..Nx+1,  j = 2..Ny+1
#
# BCs:
#   Left, right, bottom: homogeneous Dirichlet (u = 0)
#   Top y = Ly: moving Dirichlet u(x, Ly, t) = g_top(x,t)
#     BUT we do NOT drive the very corner cells to avoid incompatibility
#     with left/right boundaries:
#         - drive i=3..Nx (interior top)
#         - set i=2 and i=Nx+1 to 0 at j=Ny+1 (corner lines)
##########################################
def apply_dirichlet_u_2d(u, Nx, Ny, x, t, g_top):
    """
    Apply Dirichlet BCs to u in 2D:

    - Left, right, bottom: u = 0 (homogeneous Dirichlet).
    - Top (physical boundary j = Ny+1):
        * at i=2 and i=Nx+1 (near corners): u = 0  (match side walls)
        * at i=3..Nx: u(x, Ly, t) = g_top(x, t)
    - Top ghost layers j=Ny+2,Ny+3: second-order extrapolation in y
      using interior/boundary data.

    Parameters
    ----------
    u : 2D array, shape (Nx+4, Ny+4)
        Includes ghost cells.
    x : 1D array of x-coordinates for i=0..Nx+3
    """
    # --------------------------------------------------
    # 1) Left, right, bottom boundaries: homogeneous Dirichlet
    # --------------------------------------------------

    # Left ghosts + left boundary (i=0,1)
    u[0:2, :]       = 0.0

    # Right ghosts + right boundary (i=Nx+2,Nx+3)
    u[Nx+2:Nx+4, :] = 0.0

    # Bottom ghosts + bottom boundary (j=0,1)
    u[:, 0:2]       = 0.0

    # --------------------------------------------------
    # 2) Top physical boundary j = Ny+1
    #    We treat i=2 and i=Nx+1 as corner-aligned with side walls
    #    so we set them to zero to avoid corner incompatibility.
    #    We drive the moving boundary only on i=3..Nx.
    # --------------------------------------------------
    j_top = Ny+1

    # Corners aligned with side walls: keep u=0
    u[2,     j_top] = 0.0
    u[Nx+1,  j_top] = 0.0

    # Moving boundary on the interior portion of the top edge
    for i in range(3, Nx+1):
        u[i, j_top] = g_top(x[i], t)

    # --------------------------------------------------
    # 3) Second-order ghost cells in y above the top boundary
    #    For each i in the physical x-range (2..Nx+1), use 1D
    #    quadratic extrapolation in y:
    #       u_ghost1 = 3 u0 - 3 u1 + u2   (j = Ny+2)
    #       u_ghost2 = 6 u0 - 8 u1 + 3 u2 (j = Ny+3)
    #    where:
    #       u0 = u[i, Ny+1]   (top boundary)
    #       u1 = u[i, Ny]     (first interior below)
    #       u2 = u[i, Ny-1]   (second interior below)
    # --------------------------------------------------
    for i in range(2, Nx+2):
        u0 = u[i, Ny+1]
        u1 = u[i, Ny]
        u2 = u[i, Ny-1]

        u[i, Ny+2] = 3.0*u0 - 3.0*u1 + u2
        u[i, Ny+3] = 6.0*u0 - 8.0*u1 + 3.0*u2

    # For i=0,1 and i=Nx+2,Nx+3 we keep u=0 everywhere,
    # so we DO NOT inject the moving pattern into the ghost columns at the corners.


def apply_dirichlet_p_2d(p, Nx, Ny, x, t, g_top_t):
    """
    Apply Dirichlet-like BCs for p = u_t in 2D:

    - Left, right, bottom: p = 0 on the physical boundaries.
    - Top physical boundary j = Ny+1:
        * at i=2 and i=Nx+1: p = 0 (match side walls)
        * at i=3..Nx: p(x, Ly, t) = g_top_t(x, t)
    - Ghost cells in x and y: second-order extrapolation in the
      normal direction (same ghost strategy as u, but using p).

    This keeps p consistent with the time-dependent Dirichlet condition on u.
    """

    # --------------------------------------------------
    # 1) Left / right / bottom: p = 0 on physical boundary lines
    # --------------------------------------------------

    # Left/right vertical boundaries inside physical domain: i = 2, Nx+1
    p[2,  :]    = 0.0
    p[Nx+1, :]  = 0.0

    # Bottom horizontal boundary: j = 2
    p[:, 2]     = 0.0

    # --------------------------------------------------
    # 2) Top physical boundary j = Ny+1
    #    i=2, Nx+1: keep p=0 to match side walls;
    #    i=3..Nx: set p = time derivative of g_top.
    # --------------------------------------------------
    j_top = Ny+1

    p[2,    j_top] = 0.0
    p[Nx+1, j_top] = 0.0

    for i in range(3, Nx+1):
        p[i, j_top] = g_top_t(x[i], t)

    # --------------------------------------------------
    # 3) Second-order ghost cells in x (left/right)
    #    For each fixed j, apply quadratic extrapolation in x
    #    using boundary cells at i=2 and i=Nx+1.
    # --------------------------------------------------
    for j in range(0, Ny+4):
        # Left side: i=0,1 ghosts; boundary at i=2
        p0L = p[2, j]
        p1L = p[3, j]
        p2L = p[4, j]

        p[1, j] = 3.0*p0L - 3.0*p1L + p2L
        p[0, j] = 6.0*p0L - 8.0*p1L + 3.0*p2L

        # Right side: i=Nx+1 is boundary
        p0R = p[Nx+1, j]
        p1R = p[Nx,   j]
        p2R = p[Nx-1, j]

        p[Nx+2, j] = 3.0*p0R - 3.0*p1R + p2R
        p[Nx+3, j] = 6.0*p0R - 8.0*p1R + 3.0*p2R

    # --------------------------------------------------
    # 4) Second-order ghost cells in y (bottom/top)
    # --------------------------------------------------
    for i in range(0, Nx+4):
        # Bottom side: j=2 is physical boundary
        p0B = p[i, 2]
        p1B = p[i, 3]
        p2B = p[i, 4]

        p[i, 1] = 3.0*p0B - 3.0*p1B + p2B
        p[i, 0] = 6.0*p0B - 8.0*p1B + 3.0*p2B

        # Top side: j=Ny+1 is physical boundary (may be 0 or g_top_t)
        p0T = p[i, Ny+1]
        p1T = p[i, Ny]
        p2T = p[i, Ny-1]

        p[i, Ny+2] = 3.0*p0T - 3.0*p1T + p2T
        p[i, Ny+3] = 6.0*p0T - 8.0*p1T + 3.0*p2T


##########################################
# 2D Wave solver using dimensional splitting
# and 1D Beam‚ÄìWarming in x and y,
# with a moving upper boundary.
##########################################
def solve_wave_2d_BW_moving_top(
    c=1.0,
    Lx=1.0,
    Ly=1.0,
    Nx=100,
    Ny=100,
    Tfinal=1.0,
    CFL=0.3,
    g_top=None,
    g_top_t=None
):
    """
    2D wave equation:
        u_tt = c^2 (u_xx + u_yy)
    in first-order form with variables:
        u, p = u_t, r = c u_x, s = c u_y

    Boundary conditions:
      - Left, right, bottom: u = 0 (homogeneous Dirichlet)
      - Top: u(x, Ly, t) = g_top(x, t),
             p(x, Ly, t) = g_top_t(x, t)
        but the very corner-aligned cells (i=2 and i=Nx+1, j=Ny+1)
        are set to 0 for consistency with side walls.

    Uses dimensional splitting + 1D Beam‚ÄìWarming in x and y.
    """

    dx = Lx / Nx
    dy = Ly / Ny

    # conservative CFL choice for split scheme
    dt = CFL * min(dx, dy) / c
    Nt = int(np.ceil(Tfinal / dt))
    dt = Tfinal / Nt  # snap so that Nt*dt = Tfinal

    # Indices and sizes with 2 ghost cells each side
    Nxtot = Nx + 4
    Nytot = Ny + 4
    i_start, i_end = 2, Nx+1  # interior x range
    j_start, j_end = 2, Ny+1  # interior y range

    # Allocate arrays (all with ghost cells)
    u = np.zeros((Nxtot, Nytot))
    p = np.zeros((Nxtot, Nytot))
    r = np.zeros((Nxtot, Nytot))
    s = np.zeros((Nxtot, Nytot))

    # Coordinates for all cell centers
    x = (np.arange(Nxtot) - 2 + 0.5) * dx
    y = (np.arange(Nytot) - 2 + 0.5) * dy
    X, Y = np.meshgrid(x, y, indexing='ij')

    # Default moving top if user does not supply one:
    # A Gaussian pulse moving right with speed V.
    if g_top is None or g_top_t is None:
        A = 1.0
        beta = 200.0
        V = 1.0
        x0 = 0.25 * Lx

        def g_top(xloc, tloc):
            # Gaussian centered at x = x0 + V t
            return A * np.exp(-beta * (xloc - x0 - V*tloc)**2)

        def g_top_t(xloc, tloc):
            # time derivative of A exp(-beta (x - x0 - Vt)^2)
            xi = xloc - x0 - V*tloc
            # d/dt [exp(-beta xi^2)] = exp(-beta xi^2) * (-2 beta xi * dxi/dt)
            # dxi/dt = -V
            return A * (2.0*beta*V*xi) * np.exp(-beta * xi**2)

    # Initial condition: 2D Gaussian bump in the interior, zero initial velocity
    X_phys = X[i_start:i_end+1, j_start:j_end+1]
    Y_phys = Y[i_start:i_end+1, j_start:j_end+1]
    u[i_start:i_end+1, j_start:j_end+1] = np.exp(
        -100.0 * ((X_phys - 0.5*Lx)**2 + (Y_phys - 0.5*Ly)**2)
    )
    p[i_start:i_end+1, j_start:j_end+1] = 0.0

    # Apply initial BCs
    apply_dirichlet_u_2d(u, Nx, Ny, x, 0.0, g_top)
    apply_dirichlet_p_2d(p, Nx, Ny, x, 0.0, g_top_t)

    # Initialize r = c u_x and s = c u_y using centered differences
    r[i_start:i_end+1, j_start:j_end+1] = c * (
        u[i_start+1:i_end+2, j_start:j_end+1] - u[i_start-1:i_end, j_start:j_end+1]
    ) / (2.0*dx)

    s[i_start:i_end+1, j_start:j_end+1] = c * (
        u[i_start:i_end+1, j_start+1:j_end+2] - u[i_start:i_end+1, j_start-1:j_end]
    ) / (2.0*dy)

    # Time-stepping loop
    for n in range(Nt):
        t = n*dt

        # 1) Apply BCs for current time (moving top + homogeneous sides)
        apply_dirichlet_u_2d(u, Nx, Ny, x, t,   g_top)
        apply_dirichlet_p_2d(p, Nx, Ny, x, t,   g_top_t)

        # 2) Recompute r and s from current u (needed before each split step)
        r[i_start:i_end+1, j_start:j_end+1] = c * (
            u[i_start+1:i_end+2, j_start:j_end+1] - u[i_start-1:i_end, j_start:j_end+1]
        ) / (2.0*dx)
        s[i_start:i_end+1, j_start:j_end+1] = c * (
            u[i_start:i_end+1, j_start+1:j_end+2] - u[i_start:i_end+1, j_start-1:j_end]
        ) / (2.0*dy)

        # ============================
        # X-SWEEP: update (p,r) along x
        # ============================
        p_x_sweep = p.copy()
        r_x_sweep = r.copy()

        for j in range(j_start, j_end+1):  # loop over each y-row
            px = p[:, j]
            rx = r[:, j]

            # 1D characteristic vars in x:
            #   w_-^x = 0.5 (p + r), speed -c
            #   w_+^x = 0.5 (p - r), speed +c
            w_minus_x = 0.5 * (px + rx)
            w_plus_x  = 0.5 * (px - rx)

            # Beam‚ÄìWarming in x for each characteristic
            w_minus_x_new = BW_timestep_scalar(w_minus_x, a=-c, dx=dx, dt=dt,
                                               i_start=i_start, i_end=i_end)
            w_plus_x_new  = BW_timestep_scalar(w_plus_x,  a=+c, dx=dx, dt=dt,
                                               i_start=i_start, i_end=i_end)

            # Reconstruct p and r for this row
            p_x_sweep[:, j] = w_minus_x_new + w_plus_x_new
            r_x_sweep[:, j] = w_minus_x_new - w_plus_x_new

        # ============================
        # Y-SWEEP: update (p,s) along y
        # ============================
        p_new = p_x_sweep.copy()  # start from x-swept p
        r_new = r_x_sweep.copy()  # r is frozen in y-sweep (split)
        s_new = s.copy()

        for i in range(i_start, i_end+1):  # loop over each x-column
            py = p_x_sweep[i, :]
            sy = s[i, :]

            # 1D characteristic vars in y:
            #   w_-^y = 0.5 (p + s), speed -c (downward)
            #   w_+^y = 0.5 (p - s), speed +c (upward)
            w_minus_y = 0.5 * (py + sy)
            w_plus_y  = 0.5 * (py - sy)

            # Beam‚ÄìWarming in y for each characteristic
            w_minus_y_new = BW_timestep_scalar(w_minus_y, a=-c, dx=dy, dt=dt,
                                               i_start=j_start, i_end=j_end)
            w_plus_y_new  = BW_timestep_scalar(w_plus_y,  a=+c, dx=dy, dt=dt,
                                               i_start=j_start, i_end=j_end)

            # Reconstruct p and s for this column
            p_new[i, :] = w_minus_y_new + w_plus_y_new
            s_new[i, :] = w_minus_y_new - w_plus_y_new

        # Update p, r, s at the end of the split step
        p = p_new
        r = r_new
        s = s_new

        # 3) Update u with u_t = p (forward Euler in time)
        u[i_start:i_end+1, j_start:j_end+1] += dt * p[i_start:i_end+1, j_start:j_end+1]

    # Return physical interior and coordinates for plotting
    X_phys = X[i_start:i_end+1, j_start:j_end+1]
    Y_phys = Y[i_start:i_end+1, j_start:j_end+1]
    u_phys = u[i_start:i_end+1, j_start:j_end+1]

    return X_phys, Y_phys, u_phys


##########################################
# Demo with moving top boundary
##########################################
if __name__ == "__main__":
    c = 1.0
    Lx, Ly = 1.0, 1.0
    Nx, Ny = 120, 120

    # small final time to debug BC behavior; increase to see full evolution
    Tfinal = 0.50
    CFL = 0.010

    X, Y, U = solve_wave_2d_BW_moving_top(
        c=c, Lx=Lx, Ly=Ly, Nx=Nx, Ny=Ny,
        Tfinal=Tfinal, CFL=CFL
    )

    plt.figure(figsize=(6,5))
    plt.pcolormesh(X, Y, U, shading="auto")
    plt.colorbar(label="u(x,y,T)")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("2D Wave Equation ‚Äì Beam‚ÄìWarming + Moving Top Boundary (corner-fixed)")
    plt.tight_layout()
    plt.show()
```
